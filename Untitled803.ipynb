```python
import os
import json
import yaml
import h5py
import joblib
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

warnings.filterwarnings("ignore")

# ============================================================
# PATHS
# ============================================================

DATASET_PATH = r"C:\Users\NXTWAVE\Downloads\Chemical Production Forecasting\Production_of_Major_Chemicals_2024-25_1_0.csv"
OUTPUT_DIR = r"C:\Users\NXTWAVE\Downloads\Chemical Production Forecasting"

os.makedirs(OUTPUT_DIR, exist_ok=True)

PREFIX = "qpso_"

# ============================================================
# LOAD DATA
# ============================================================

df = pd.read_csv(DATASET_PATH)
df.columns = df.columns.str.strip()

print("Dataset Loaded Successfully")
print("Shape:", df.shape)
print(df.head())

# ============================================================
# CLEAN DATA
# ============================================================

df = df.replace(["-", "NA", "N/A", "na", "n/a", "", " "], np.nan)

for col in df.columns:
    if df[col].dtype == "object":
        cleaned = df[col].astype(str).str.replace(",", "", regex=False).str.strip()
        df[col] = pd.to_numeric(cleaned, errors="ignore")

for col in df.columns:
    if df[col].dtype == "object":
        df[col] = df[col].fillna("Unknown")
    else:
        df[col] = df[col].fillna(df[col].median())

# ============================================================
# TARGET SELECTION
# ============================================================

numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()

if len(numeric_cols) == 0:
    raise ValueError("No numeric column found for prediction.")

target_col = numeric_cols[-1]

print("Target Column Selected:", target_col)

X = df.drop(columns=[target_col])
y = df[target_col]

# ============================================================
# ENCODING
# ============================================================

label_encoders = {}

for col in X.columns:
    if X[col].dtype == "object":
        le = LabelEncoder()
        X[col] = le.fit_transform(X[col].astype(str))
        label_encoders[col] = le

feature_names = list(X.columns)

# ============================================================
# SCALING
# ============================================================

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ============================================================
# TRAIN TEST SPLIT
# ============================================================

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.2,
    random_state=42
)

# ============================================================
# QPSO FEATURE SELECTION
# Quantum Particle Swarm Optimization
# ============================================================

def evaluate_solution(solution, X_train, X_test, y_train, y_test):
    selected_features = np.where(solution == 1)[0]

    if len(selected_features) == 0:
        return -9999

    model = RandomForestRegressor(
        n_estimators=100,
        max_depth=10,
        random_state=42
    )

    model.fit(X_train[:, selected_features], y_train)
    pred = model.predict(X_test[:, selected_features])

    return r2_score(y_test, pred)


def sigmoid(x):
    x = np.clip(x, -500, 500)
    return 1 / (1 + np.exp(-x))


def qpso_feature_selection(
    X_train,
    X_test,
    y_train,
    y_test,
    num_features,
    num_particles=25,
    iterations=30,
    beta_start=1.0,
    beta_end=0.5
):
    positions = np.random.rand(num_particles, num_features)
    binary_positions = (positions > 0.5).astype(int)

    for i in range(num_particles):
        if np.sum(binary_positions[i]) == 0:
            binary_positions[i, np.random.randint(0, num_features)] = 1

    personal_best_positions = positions.copy()
    personal_best_binary = binary_positions.copy()

    personal_best_scores = np.array([
        evaluate_solution(individual, X_train, X_test, y_train, y_test)
        for individual in personal_best_binary
    ])

    best_index = np.argmax(personal_best_scores)

    global_best_position = personal_best_positions[best_index].copy()
    global_best_binary = personal_best_binary[best_index].copy()
    global_best_score = personal_best_scores[best_index]

    progress = []

    for iteration in range(iterations):
        beta = beta_start - ((beta_start - beta_end) * iteration / iterations)

        mbest = np.mean(personal_best_positions, axis=0)

        for i in range(num_particles):
            phi = np.random.rand(num_features)

            local_attractor = (
                phi * personal_best_positions[i]
                + (1 - phi) * global_best_position
            )

            u = np.random.rand(num_features)
            direction = np.where(np.random.rand(num_features) < 0.5, 1, -1)

            new_position = local_attractor + (
                direction * beta * np.abs(mbest - positions[i]) * np.log(1 / u)
            )

            positions[i] = np.clip(new_position, 0, 1)

            probabilities = sigmoid(10 * (positions[i] - 0.5))
            binary_positions[i] = (
                np.random.rand(num_features) < probabilities
            ).astype(int)

            if np.sum(binary_positions[i]) == 0:
                binary_positions[i, np.random.randint(0, num_features)] = 1

            score = evaluate_solution(
                binary_positions[i],
                X_train,
                X_test,
                y_train,
                y_test
            )

            if score > personal_best_scores[i]:
                personal_best_scores[i] = score
                personal_best_positions[i] = positions[i].copy()
                personal_best_binary[i] = binary_positions[i].copy()

                if score > global_best_score:
                    global_best_score = score
                    global_best_position = positions[i].copy()
                    global_best_binary = binary_positions[i].copy()

        progress.append(global_best_score)

        print(
            f"Iteration {iteration + 1}/{iterations} | "
            f"Best R2 Score: {global_best_score:.4f} | "
            f"Selected Features: {np.sum(global_best_binary)}"
        )

    selected_indices = np.where(global_best_binary == 1)[0]

    return selected_indices, global_best_score, progress


selected_indices, qpso_best_score, qpso_progress = qpso_feature_selection(
    X_train,
    X_test,
    y_train,
    y_test,
    num_features=X_train.shape[1],
    num_particles=25,
    iterations=30,
    beta_start=1.0,
    beta_end=0.5
)

selected_features = [feature_names[i] for i in selected_indices]

print("\nQPSO Selected Features:")
print(selected_features)

X_train_qpso = X_train[:, selected_indices]
X_test_qpso = X_test[:, selected_indices]

# ============================================================
# MODEL 1: RANDOM FOREST
# ============================================================

rf_model = RandomForestRegressor(
    n_estimators=200,
    max_depth=10,
    random_state=42
)

rf_model.fit(X_train_qpso, y_train)
rf_pred = rf_model.predict(X_test_qpso)

rf_mae = mean_absolute_error(y_test, rf_pred)
rf_mse = mean_squared_error(y_test, rf_pred)
rf_rmse = np.sqrt(rf_mse)
rf_r2 = r2_score(y_test, rf_pred)

# ============================================================
# MODEL 2: MLP REGRESSOR
# TensorFlow removed to avoid DLL paging file error
# ============================================================

mlp_model = MLPRegressor(
    hidden_layer_sizes=(128, 64, 32),
    activation="relu",
    solver="adam",
    max_iter=1000,
    random_state=42,
    early_stopping=True
)

mlp_model.fit(X_train_qpso, y_train)
mlp_pred = mlp_model.predict(X_test_qpso)

mlp_mae = mean_absolute_error(y_test, mlp_pred)
mlp_mse = mean_squared_error(y_test, mlp_pred)
mlp_rmse = np.sqrt(mlp_mse)
mlp_r2 = r2_score(y_test, mlp_pred)

# ============================================================
# SAVE MODELS
# ============================================================

rf_pkl_path = os.path.join(OUTPUT_DIR, PREFIX + "random_forest_model.pkl")
mlp_pkl_path = os.path.join(OUTPUT_DIR, PREFIX + "mlp_model.pkl")
scaler_path = os.path.join(OUTPUT_DIR, PREFIX + "scaler.pkl")
encoder_path = os.path.join(OUTPUT_DIR, PREFIX + "label_encoders.pkl")
selected_features_path = os.path.join(OUTPUT_DIR, PREFIX + "selected_features.pkl")
h5_path = os.path.join(OUTPUT_DIR, PREFIX + "model_data.h5")

joblib.dump(rf_model, rf_pkl_path)
joblib.dump(mlp_model, mlp_pkl_path)
joblib.dump(scaler, scaler_path)
joblib.dump(label_encoders, encoder_path)
joblib.dump(selected_features, selected_features_path)

with h5py.File(h5_path, "w") as h5:
    h5.create_dataset("selected_feature_indices", data=selected_indices)
    h5.create_dataset("rf_predictions", data=rf_pred)
    h5.create_dataset("mlp_predictions", data=mlp_pred)
    h5.create_dataset("actual_values", data=y_test.values)
    h5.attrs["project_name"] = "Chemical Production Forecasting using QPSO"
    h5.attrs["target_column"] = target_col
    h5.attrs["qpso_best_score"] = qpso_best_score

# ============================================================
# SAVE CSV FILES
# ============================================================

prediction_df = pd.DataFrame({
    "Actual": y_test.values,
    "QPSO_Random_Forest_Predicted": rf_pred,
    "QPSO_MLP_Predicted": mlp_pred
})

prediction_csv_path = os.path.join(OUTPUT_DIR, PREFIX + "prediction.csv")
prediction_df.to_csv(prediction_csv_path, index=False)

results_df = pd.DataFrame({
    "Model": ["QPSO + Random Forest", "QPSO + MLP Regressor"],
    "MAE": [rf_mae, mlp_mae],
    "MSE": [rf_mse, mlp_mse],
    "RMSE": [rf_rmse, mlp_rmse],
    "R2_Score": [rf_r2, mlp_r2],
    "Accuracy_Percentage": [rf_r2 * 100, mlp_r2 * 100],
    "Selected_Features_Count": [len(selected_features), len(selected_features)]
})

result_csv_path = os.path.join(OUTPUT_DIR, PREFIX + "result.csv")
results_df.to_csv(result_csv_path, index=False)

print("\nQPSO Result CSV:")
print(results_df)

# ============================================================
# JSON REPORT
# ============================================================

json_report = {
    "project_name": "Chemical Production Forecasting System using QPSO",
    "author": "Sagnik Patra",
    "algorithm": "Quantum Particle Swarm Optimization Feature Selection",
    "note": "TensorFlow removed. MLPRegressor used to avoid DLL paging file error.",
    "target_column": target_col,
    "rows": int(df.shape[0]),
    "columns": int(df.shape[1]),
    "total_features": int(len(feature_names)),
    "selected_features_count": int(len(selected_features)),
    "selected_features": selected_features,
    "qpso_best_score": float(qpso_best_score),
    "random_forest": {
        "mae": float(rf_mae),
        "mse": float(rf_mse),
        "rmse": float(rf_rmse),
        "r2_score": float(rf_r2),
        "accuracy_percentage": float(rf_r2 * 100)
    },
    "mlp_regressor": {
        "mae": float(mlp_mae),
        "mse": float(mlp_mse),
        "rmse": float(mlp_rmse),
        "r2_score": float(mlp_r2),
        "accuracy_percentage": float(mlp_r2 * 100)
    }
}

json_path = os.path.join(OUTPUT_DIR, PREFIX + "report.json")

with open(json_path, "w") as f:
    json.dump(json_report, f, indent=4)

# ============================================================
# YAML CONFIG
# ============================================================

yaml_config = {
    "project": {
        "name": "Chemical Production Forecasting System using QPSO",
        "author": "Sagnik Patra"
    },
    "note": "TensorFlow removed due to DLL paging file error. MLPRegressor used instead.",
    "paths": {
        "dataset_path": DATASET_PATH,
        "output_directory": OUTPUT_DIR,
        "h5_file": h5_path,
        "rf_pkl_model": rf_pkl_path,
        "mlp_pkl_model": mlp_pkl_path,
        "prediction_csv": prediction_csv_path,
        "result_csv": result_csv_path,
        "json_report": json_path
    },
    "data": {
        "target_column": target_col,
        "rows": int(df.shape[0]),
        "columns": int(df.shape[1]),
        "all_features": feature_names,
        "selected_features": selected_features
    },
    "qpso_parameters": {
        "num_particles": 25,
        "iterations": 30,
        "beta_start": 1.0,
        "beta_end": 0.5
    },
    "models": {
        "random_forest": {
            "n_estimators": 200,
            "max_depth": 10,
            "random_state": 42
        },
        "mlp_regressor": {
            "hidden_layer_sizes": [128, 64, 32],
            "activation": "relu",
            "solver": "adam",
            "max_iter": 1000,
            "early_stopping": True
        }
    }
}

yaml_path = os.path.join(OUTPUT_DIR, PREFIX + "config.yaml")

with open(yaml_path, "w") as f:
    yaml.dump(yaml_config, f, sort_keys=False)

# ============================================================
# GRAPHS
# ============================================================

qpso_progress_graph_path = os.path.join(
    OUTPUT_DIR,
    PREFIX + "optimization_progress_graph.png"
)

plt.figure(figsize=(9, 5))
plt.plot(qpso_progress, marker="o")
plt.title("QPSO Optimization Progress Graph")
plt.xlabel("Iteration")
plt.ylabel("Best R2 Score")
plt.tight_layout()
plt.savefig(qpso_progress_graph_path, dpi=300)
plt.show()

accuracy_graph_path = os.path.join(OUTPUT_DIR, PREFIX + "accuracy_graph.png")

plt.figure(figsize=(8, 5))
sns.barplot(x="Model", y="Accuracy_Percentage", data=results_df)
plt.title("QPSO Model Accuracy Comparison")
plt.xlabel("Model")
plt.ylabel("Accuracy Percentage")
plt.xticks(rotation=15)
plt.tight_layout()
plt.savefig(accuracy_graph_path, dpi=300)
plt.show()

result_graph_path = os.path.join(OUTPUT_DIR, PREFIX + "result_graph.png")

plt.figure(figsize=(8, 5))
sns.barplot(x="Model", y="RMSE", data=results_df)
plt.title("QPSO Model RMSE Result Comparison")
plt.xlabel("Model")
plt.ylabel("RMSE")
plt.xticks(rotation=15)
plt.tight_layout()
plt.savefig(result_graph_path, dpi=300)
plt.show()

comparison_graph_path = os.path.join(OUTPUT_DIR, PREFIX + "comparison_graph.png")

plt.figure(figsize=(10, 6))
plt.plot(y_test.values, label="Actual", marker="o")
plt.plot(rf_pred, label="QPSO Random Forest Predicted", marker="x")
plt.plot(mlp_pred, label="QPSO MLP Predicted", marker="s")
plt.title("QPSO Actual vs Predicted Comparison")
plt.xlabel("Test Sample Index")
plt.ylabel(target_col)
plt.legend()
plt.tight_layout()
plt.savefig(comparison_graph_path, dpi=300)
plt.show()

prediction_graph_path = os.path.join(OUTPUT_DIR, PREFIX + "prediction_graph.png")

plt.figure(figsize=(10, 6))
plt.scatter(y_test.values, rf_pred, label="QPSO Random Forest")
plt.scatter(y_test.values, mlp_pred, label="QPSO MLP")
plt.plot(
    [y_test.min(), y_test.max()],
    [y_test.min(), y_test.max()],
    linestyle="--",
    label="Perfect Prediction Line"
)
plt.title("QPSO Prediction Graph")
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.legend()
plt.tight_layout()
plt.savefig(prediction_graph_path, dpi=300)
plt.show()

heatmap_path = os.path.join(OUTPUT_DIR, PREFIX + "heatmap.png")

encoded_df = df.copy()

for col in encoded_df.columns:
    if encoded_df[col].dtype == "object":
        le = LabelEncoder()
        encoded_df[col] = le.fit_transform(encoded_df[col].astype(str))

plt.figure(figsize=(12, 8))
sns.heatmap(encoded_df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("QPSO Feature Correlation Heatmap")
plt.tight_layout()
plt.savefig(heatmap_path, dpi=300)
plt.show()

loss_graph_path = os.path.join(OUTPUT_DIR, PREFIX + "training_loss_graph.png")

plt.figure(figsize=(8, 5))
plt.plot(mlp_model.loss_curve_, label="MLP Training Loss")
plt.title("QPSO MLP Training Loss Graph")
plt.xlabel("Iteration")
plt.ylabel("Loss")
plt.legend()
plt.tight_layout()
plt.savefig(loss_graph_path, dpi=300)
plt.show()

# ============================================================
# FINAL OUTPUT
# ============================================================

print("\nAll QPSO files saved successfully!")

print("RF PKL Model:", rf_pkl_path)
print("MLP PKL Model:", mlp_pkl_path)
print("H5 File:", h5_path)
print("Scaler:", scaler_path)
print("Encoders:", encoder_path)
print("Selected Features:", selected_features_path)
print("YAML Config:", yaml_path)
print("JSON Report:", json_path)
print("Result CSV:", result_csv_path)
print("Prediction CSV:", prediction_csv_path)
print("Optimization Graph:", qpso_progress_graph_path)
print("Accuracy Graph:", accuracy_graph_path)
print("Result Graph:", result_graph_path)
print("Comparison Graph:", comparison_graph_path)
print("Prediction Graph:", prediction_graph_path)
print("Heatmap:", heatmap_path)
print("Training Loss Graph:", loss_graph_path)
```
