```python
import os
import json
import yaml
import joblib
import warnings
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping

warnings.filterwarnings("ignore")

# ============================================================
# PATHS
# ============================================================

DATASET_PATH = r"C:\Users\NXTWAVE\Downloads\Chemical Production Forecasting\Production_of_Major_Chemicals_2024-25_1_0.csv"

OUTPUT_DIR = r"C:\Users\NXTWAVE\Downloads\Chemical Production Forecasting"

os.makedirs(OUTPUT_DIR, exist_ok=True)

PREFIX = "chemical_"

# ============================================================
# LOAD DATA
# ============================================================

df = pd.read_csv(DATASET_PATH)

print("Dataset Loaded Successfully")
print("Shape:", df.shape)
print(df.head())

# ============================================================
# BASIC CLEANING
# ============================================================

df.columns = df.columns.str.strip()

df = df.replace(["-", "NA", "N/A", "na", "n/a", ""], np.nan)

# Convert possible numeric columns
for col in df.columns:
    if df[col].dtype == "object":
        try:
            df[col] = (
                df[col]
                .astype(str)
                .str.replace(",", "", regex=False)
                .str.strip()
            )
            df[col] = pd.to_numeric(df[col])
        except:
            pass

# Fill missing values
for col in df.columns:
    if df[col].dtype == "object":
        df[col] = df[col].fillna("Unknown")
    else:
        df[col] = df[col].fillna(df[col].median())

# ============================================================
# AUTO DETECT TARGET COLUMN
# ============================================================

numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()

if len(numeric_cols) == 0:
    raise ValueError("No numeric column found for prediction.")

target_col = numeric_cols[-1]

print("Target Column Selected:", target_col)

X = df.drop(columns=[target_col])
y = df[target_col]

# ============================================================
# ENCODING CATEGORICAL COLUMNS
# ============================================================

label_encoders = {}

for col in X.columns:
    if X[col].dtype == "object":
        le = LabelEncoder()
        X[col] = le.fit_transform(X[col].astype(str))
        label_encoders[col] = le

# ============================================================
# SCALING
# ============================================================

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ============================================================
# TRAIN TEST SPLIT
# ============================================================

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.2,
    random_state=42
)

# ============================================================
# RANDOM FOREST MODEL
# ============================================================

rf_model = RandomForestRegressor(
    n_estimators=200,
    random_state=42,
    max_depth=10
)

rf_model.fit(X_train, y_train)

rf_pred = rf_model.predict(X_test)

rf_mae = mean_absolute_error(y_test, rf_pred)
rf_mse = mean_squared_error(y_test, rf_pred)
rf_rmse = np.sqrt(rf_mse)
rf_r2 = r2_score(y_test, rf_pred)

print("\nRandom Forest Results")
print("MAE:", rf_mae)
print("RMSE:", rf_rmse)
print("R2 Score:", rf_r2)

# ============================================================
# DEEP LEARNING H5 MODEL
# ============================================================

dl_model = Sequential([
    Dense(128, activation="relu", input_shape=(X_train.shape[1],)),
    Dropout(0.2),
    Dense(64, activation="relu"),
    Dropout(0.2),
    Dense(32, activation="relu"),
    Dense(1)
])

dl_model.compile(
    optimizer="adam",
    loss="mse",
    metrics=["mae"]
)

early_stop = EarlyStopping(
    monitor="val_loss",
    patience=20,
    restore_best_weights=True
)

history = dl_model.fit(
    X_train,
    y_train,
    validation_split=0.2,
    epochs=200,
    batch_size=8,
    callbacks=[early_stop],
    verbose=1
)

dl_pred = dl_model.predict(X_test).flatten()

dl_mae = mean_absolute_error(y_test, dl_pred)
dl_mse = mean_squared_error(y_test, dl_pred)
dl_rmse = np.sqrt(dl_mse)
dl_r2 = r2_score(y_test, dl_pred)

print("\nDeep Learning Results")
print("MAE:", dl_mae)
print("RMSE:", dl_rmse)
print("R2 Score:", dl_r2)

# ============================================================
# SAVE MODEL FILES
# ============================================================

h5_path = os.path.join(OUTPUT_DIR, PREFIX + "model.h5")
pkl_model_path = os.path.join(OUTPUT_DIR, PREFIX + "random_forest_model.pkl")
scaler_path = os.path.join(OUTPUT_DIR, PREFIX + "scaler.pkl")
encoder_path = os.path.join(OUTPUT_DIR, PREFIX + "label_encoders.pkl")

dl_model.save(h5_path)
joblib.dump(rf_model, pkl_model_path)
joblib.dump(scaler, scaler_path)
joblib.dump(label_encoders, encoder_path)

# ============================================================
# SAVE PREDICTIONS CSV
# ============================================================

prediction_df = pd.DataFrame({
    "Actual": y_test.values,
    "Random_Forest_Predicted": rf_pred,
    "Deep_Learning_Predicted": dl_pred
})

prediction_csv_path = os.path.join(OUTPUT_DIR, PREFIX + "predictions.csv")
prediction_df.to_csv(prediction_csv_path, index=False)

# ============================================================
# SAVE RESULTS CSV
# ============================================================

results_df = pd.DataFrame({
    "Model": ["Random Forest", "Deep Learning H5"],
    "MAE": [rf_mae, dl_mae],
    "MSE": [rf_mse, dl_mse],
    "RMSE": [rf_rmse, dl_rmse],
    "R2_Score": [rf_r2, dl_r2]
})

results_csv_path = os.path.join(OUTPUT_DIR, PREFIX + "results.csv")
results_df.to_csv(results_csv_path, index=False)

# ============================================================
# SAVE JSON REPORT
# ============================================================

json_report = {
    "project_name": "Chemical Production Forecasting System",
    "dataset_path": DATASET_PATH,
    "output_directory": OUTPUT_DIR,
    "target_column": target_col,
    "total_rows": int(df.shape[0]),
    "total_columns": int(df.shape[1]),
    "features_used": list(X.columns),
    "models": {
        "random_forest": {
            "mae": float(rf_mae),
            "mse": float(rf_mse),
            "rmse": float(rf_rmse),
            "r2_score": float(rf_r2)
        },
        "deep_learning_h5": {
            "mae": float(dl_mae),
            "mse": float(dl_mse),
            "rmse": float(dl_rmse),
            "r2_score": float(dl_r2)
        }
    },
    "saved_files": {
        "h5_model": h5_path,
        "pkl_model": pkl_model_path,
        "scaler": scaler_path,
        "label_encoders": encoder_path,
        "prediction_csv": prediction_csv_path,
        "results_csv": results_csv_path
    }
}

json_path = os.path.join(OUTPUT_DIR, PREFIX + "report.json")

with open(json_path, "w") as f:
    json.dump(json_report, f, indent=4)

# ============================================================
# SAVE YAML CONFIG
# ============================================================

yaml_config = {
    "project": {
        "name": "Chemical Production Forecasting System",
        "author": "Sagnik Patra"
    },
    "paths": {
        "dataset": DATASET_PATH,
        "output_directory": OUTPUT_DIR,
        "h5_model": h5_path,
        "pkl_model": pkl_model_path,
        "json_report": json_path
    },
    "data": {
        "target_column": target_col,
        "features": list(X.columns),
        "rows": int(df.shape[0]),
        "columns": int(df.shape[1])
    },
    "models": {
        "random_forest": {
            "n_estimators": 200,
            "max_depth": 10,
            "random_state": 42
        },
        "deep_learning": {
            "layers": [128, 64, 32, 1],
            "activation": "relu",
            "optimizer": "adam",
            "loss": "mse",
            "epochs": 200,
            "batch_size": 8
        }
    }
}

yaml_path = os.path.join(OUTPUT_DIR, PREFIX + "config.yaml")

with open(yaml_path, "w") as f:
    yaml.dump(yaml_config, f, sort_keys=False)

# ============================================================
# FINAL OUTPUT
# ============================================================

print("\nAll files saved successfully!")

print("\nSaved Files:")
print("H5 Model:", h5_path)
print("PKL Model:", pkl_model_path)
print("Scaler PKL:", scaler_path)
print("Encoder PKL:", encoder_path)
print("Prediction CSV:", prediction_csv_path)
print("Results CSV:", results_csv_path)
print("JSON Report:", json_path)
print("YAML Config:", yaml_path)
```
