```python
import os
import json
import yaml
import h5py
import joblib
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

warnings.filterwarnings("ignore")

# ============================================================
# PATHS
# ============================================================

DATASET_PATH = r"C:\Users\NXTWAVE\Downloads\Chemical Production Forecasting\Production_of_Major_Chemicals_2024-25_1_0.csv"
OUTPUT_DIR = r"C:\Users\NXTWAVE\Downloads\Chemical Production Forecasting"

os.makedirs(OUTPUT_DIR, exist_ok=True)

PREFIX = "bis_"

# ============================================================
# LOAD DATA
# ============================================================

df = pd.read_csv(DATASET_PATH)
df.columns = df.columns.str.strip()

print("Dataset Loaded Successfully")
print("Shape:", df.shape)
print(df.head())

# ============================================================
# CLEAN DATA
# ============================================================

df = df.replace(["-", "NA", "N/A", "na", "n/a", "", " "], np.nan)

for col in df.columns:
    if df[col].dtype == "object":
        cleaned = df[col].astype(str).str.replace(",", "", regex=False).str.strip()
        df[col] = pd.to_numeric(cleaned, errors="ignore")

for col in df.columns:
    if df[col].dtype == "object":
        df[col] = df[col].fillna("Unknown")
    else:
        df[col] = df[col].fillna(df[col].median())

# ============================================================
# TARGET SELECTION
# ============================================================

numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()

if len(numeric_cols) == 0:
    raise ValueError("No numeric column found for prediction.")

target_col = numeric_cols[-1]
print("Target Column Selected:", target_col)

X = df.drop(columns=[target_col])
y = df[target_col]

# ============================================================
# ENCODING
# ============================================================

label_encoders = {}

for col in X.columns:
    if X[col].dtype == "object":
        le = LabelEncoder()
        X[col] = le.fit_transform(X[col].astype(str))
        label_encoders[col] = le

feature_names = list(X.columns)

# ============================================================
# SCALING
# ============================================================

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ============================================================
# TRAIN TEST SPLIT
# ============================================================

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.2,
    random_state=42
)

# ============================================================
# FITNESS FUNCTION
# ============================================================

def evaluate_solution(solution, X_train, X_test, y_train, y_test):
    selected_features = np.where(solution == 1)[0]

    if len(selected_features) == 0:
        return -9999

    model = RandomForestRegressor(
        n_estimators=100,
        max_depth=10,
        random_state=42
    )

    model.fit(X_train[:, selected_features], y_train)
    pred = model.predict(X_test[:, selected_features])

    return r2_score(y_test, pred)

# ============================================================
# BA + AIS HYBRID FEATURE SELECTION
# Bat Algorithm + Artificial Immune System
# ============================================================

def sigmoid(x):
    x = np.clip(x, -500, 500)
    return 1 / (1 + np.exp(-x))


def ba_ais_feature_selection(
    X_train,
    X_test,
    y_train,
    y_test,
    num_features,
    population_size=25,
    generations=30,
    frequency_min=0.0,
    frequency_max=2.0,
    loudness=0.5,
    pulse_rate=0.5,
    clone_factor=4,
    mutation_rate=0.1
):
    bats = np.random.randint(0, 2, size=(population_size, num_features))
    velocities = np.random.uniform(-1, 1, size=(population_size, num_features))
    frequencies = np.zeros(population_size)

    for i in range(population_size):
        if np.sum(bats[i]) == 0:
            bats[i, np.random.randint(0, num_features)] = 1

    scores = np.array([
        evaluate_solution(bat, X_train, X_test, y_train, y_test)
        for bat in bats
    ])

    best_index = np.argmax(scores)
    best_bat = bats[best_index].copy()
    best_score = scores[best_index]

    progress = []

    for generation in range(generations):

        # ====================================================
        # BAT ALGORITHM UPDATE
        # ====================================================

        for i in range(population_size):
            frequencies[i] = frequency_min + (
                frequency_max - frequency_min
            ) * np.random.rand()

            velocities[i] = velocities[i] + (bats[i] - best_bat) * frequencies[i]

            probabilities = sigmoid(velocities[i])
            new_bat = (np.random.rand(num_features) < probabilities).astype(int)

            if np.random.rand() > pulse_rate:
                local_search = best_bat.copy()
                mutation_mask = np.random.rand(num_features) < mutation_rate
                local_search[mutation_mask] = 1 - local_search[mutation_mask]
                new_bat = local_search

            if np.sum(new_bat) == 0:
                new_bat[np.random.randint(0, num_features)] = 1

            new_score = evaluate_solution(
                new_bat,
                X_train,
                X_test,
                y_train,
                y_test
            )

            if new_score > scores[i] and np.random.rand() < loudness:
                bats[i] = new_bat.copy()
                scores[i] = new_score

            if scores[i] > best_score:
                best_score = scores[i]
                best_bat = bats[i].copy()

        # ====================================================
        # AIS CLONING + MUTATION ON BEST BATS
        # ====================================================

        sorted_indices = np.argsort(scores)[::-1]
        elite_count = max(3, population_size // 5)
        elites = bats[sorted_indices[:elite_count]]

        clones = []

        for elite in elites:
            for _ in range(clone_factor):
                clone = elite.copy()

                adaptive_mutation = mutation_rate * np.random.uniform(0.5, 1.5)
                mutation_mask = np.random.rand(num_features) < adaptive_mutation
                clone[mutation_mask] = 1 - clone[mutation_mask]

                if np.sum(clone) == 0:
                    clone[np.random.randint(0, num_features)] = 1

                clones.append(clone)

        clone_scores = np.array([
            evaluate_solution(clone, X_train, X_test, y_train, y_test)
            for clone in clones
        ])

        if len(clone_scores) > 0:
            best_clone_index = np.argmax(clone_scores)

            if clone_scores[best_clone_index] > best_score:
                best_score = clone_scores[best_clone_index]
                best_bat = clones[best_clone_index].copy()

        # Replace weakest bats with best clones
        if len(clones) > 0:
            worst_indices = np.argsort(scores)[:min(len(clones), elite_count)]
            best_clone_indices = np.argsort(clone_scores)[::-1][:len(worst_indices)]

            for wi, ci in zip(worst_indices, best_clone_indices):
                bats[wi] = clones[ci].copy()
                scores[wi] = clone_scores[ci]

        progress.append(best_score)

        print(
            f"Generation {generation + 1}/{generations} | "
            f"Best R2 Score: {best_score:.4f} | "
            f"Selected Features: {np.sum(best_bat)}"
        )

    selected_indices = np.where(best_bat == 1)[0]

    return selected_indices, best_score, progress


selected_indices, bis_best_score, bis_progress = ba_ais_feature_selection(
    X_train,
    X_test,
    y_train,
    y_test,
    num_features=X_train.shape[1],
    population_size=25,
    generations=30,
    frequency_min=0.0,
    frequency_max=2.0,
    loudness=0.5,
    pulse_rate=0.5,
    clone_factor=4,
    mutation_rate=0.1
)

selected_features = [feature_names[i] for i in selected_indices]

print("\nBIS Selected Features:")
print(selected_features)

X_train_bis = X_train[:, selected_indices]
X_test_bis = X_test[:, selected_indices]

# ============================================================
# MODEL 1: RANDOM FOREST
# ============================================================

rf_model = RandomForestRegressor(
    n_estimators=200,
    max_depth=10,
    random_state=42
)

rf_model.fit(X_train_bis, y_train)
rf_pred = rf_model.predict(X_test_bis)

rf_mae = mean_absolute_error(y_test, rf_pred)
rf_mse = mean_squared_error(y_test, rf_pred)
rf_rmse = np.sqrt(rf_mse)
rf_r2 = r2_score(y_test, rf_pred)

# ============================================================
# MODEL 2: MLP REGRESSOR
# TensorFlow removed to avoid DLL paging file error
# ============================================================

mlp_model = MLPRegressor(
    hidden_layer_sizes=(128, 64, 32),
    activation="relu",
    solver="adam",
    max_iter=1000,
    random_state=42,
    early_stopping=True
)

mlp_model.fit(X_train_bis, y_train)
mlp_pred = mlp_model.predict(X_test_bis)

mlp_mae = mean_absolute_error(y_test, mlp_pred)
mlp_mse = mean_squared_error(y_test, mlp_pred)
mlp_rmse = np.sqrt(mlp_mse)
mlp_r2 = r2_score(y_test, mlp_pred)

# ============================================================
# SAVE MODELS
# ============================================================

rf_pkl_path = os.path.join(OUTPUT_DIR, PREFIX + "random_forest_model.pkl")
mlp_pkl_path = os.path.join(OUTPUT_DIR, PREFIX + "mlp_model.pkl")
scaler_path = os.path.join(OUTPUT_DIR, PREFIX + "scaler.pkl")
encoder_path = os.path.join(OUTPUT_DIR, PREFIX + "label_encoders.pkl")
selected_features_path = os.path.join(OUTPUT_DIR, PREFIX + "selected_features.pkl")
h5_path = os.path.join(OUTPUT_DIR, PREFIX + "model_data.h5")

joblib.dump(rf_model, rf_pkl_path)
joblib.dump(mlp_model, mlp_pkl_path)
joblib.dump(scaler, scaler_path)
joblib.dump(label_encoders, encoder_path)
joblib.dump(selected_features, selected_features_path)

with h5py.File(h5_path, "w") as h5:
    h5.create_dataset("selected_feature_indices", data=selected_indices)
    h5.create_dataset("rf_predictions", data=rf_pred)
    h5.create_dataset("mlp_predictions", data=mlp_pred)
    h5.create_dataset("actual_values", data=y_test.values)
    h5.attrs["project_name"] = "Chemical Production Forecasting using BA + AIS"
    h5.attrs["target_column"] = target_col
    h5.attrs["bis_best_score"] = bis_best_score

# ============================================================
# SAVE CSV FILES
# ============================================================

prediction_df = pd.DataFrame({
    "Actual": y_test.values,
    "BIS_Random_Forest_Predicted": rf_pred,
    "BIS_MLP_Predicted": mlp_pred
})

prediction_csv_path = os.path.join(OUTPUT_DIR, PREFIX + "prediction.csv")
prediction_df.to_csv(prediction_csv_path, index=False)

results_df = pd.DataFrame({
    "Model": ["BA + AIS + Random Forest", "BA + AIS + MLP Regressor"],
    "MAE": [rf_mae, mlp_mae],
    "MSE": [rf_mse, mlp_mse],
    "RMSE": [rf_rmse, mlp_rmse],
    "R2_Score": [rf_r2, mlp_r2],
    "Accuracy_Percentage": [rf_r2 * 100, mlp_r2 * 100],
    "Selected_Features_Count": [len(selected_features), len(selected_features)]
})

result_csv_path = os.path.join(OUTPUT_DIR, PREFIX + "result.csv")
results_df.to_csv(result_csv_path, index=False)

print("\nBIS Result CSV:")
print(results_df)

# ============================================================
# JSON REPORT
# ============================================================

json_report = {
    "project_name": "Chemical Production Forecasting System using BA + AIS",
    "author": "Sagnik Patra",
    "algorithm": "Hybrid Bat Algorithm + Artificial Immune System",
    "prefix": PREFIX,
    "note": "TensorFlow removed. MLPRegressor used to avoid DLL paging file error.",
    "target_column": target_col,
    "rows": int(df.shape[0]),
    "columns": int(df.shape[1]),
    "total_features": int(len(feature_names)),
    "selected_features_count": int(len(selected_features)),
    "selected_features": selected_features,
    "bis_best_score": float(bis_best_score),
    "random_forest": {
        "mae": float(rf_mae),
        "mse": float(rf_mse),
        "rmse": float(rf_rmse),
        "r2_score": float(rf_r2),
        "accuracy_percentage": float(rf_r2 * 100)
    },
    "mlp_regressor": {
        "mae": float(mlp_mae),
        "mse": float(mlp_mse),
        "rmse": float(mlp_rmse),
        "r2_score": float(mlp_r2),
        "accuracy_percentage": float(mlp_r2 * 100)
    }
}

json_path = os.path.join(OUTPUT_DIR, PREFIX + "report.json")

with open(json_path, "w") as f:
    json.dump(json_report, f, indent=4)

# ============================================================
# YAML CONFIG
# ============================================================

yaml_config = {
    "project": {
        "name": "Chemical Production Forecasting System using BA + AIS",
        "author": "Sagnik Patra"
    },
    "prefix": PREFIX,
    "note": "TensorFlow removed due to DLL paging file error. MLPRegressor used instead.",
    "paths": {
        "dataset_path": DATASET_PATH,
        "output_directory": OUTPUT_DIR,
        "h5_file": h5_path,
        "rf_pkl_model": rf_pkl_path,
        "mlp_pkl_model": mlp_pkl_path,
        "prediction_csv": prediction_csv_path,
        "result_csv": result_csv_path,
        "json_report": json_path
    },
    "data": {
        "target_column": target_col,
        "rows": int(df.shape[0]),
        "columns": int(df.shape[1]),
        "all_features": feature_names,
        "selected_features": selected_features
    },
    "hybrid_parameters": {
        "algorithm": "BA + AIS",
        "population_size": 25,
        "generations": 30,
        "frequency_min": 0.0,
        "frequency_max": 2.0,
        "loudness": 0.5,
        "pulse_rate": 0.5,
        "clone_factor": 4,
        "mutation_rate": 0.1
    },
    "models": {
        "random_forest": {
            "n_estimators": 200,
            "max_depth": 10,
            "random_state": 42
        },
        "mlp_regressor": {
            "hidden_layer_sizes": [128, 64, 32],
            "activation": "relu",
            "solver": "adam",
            "max_iter": 1000,
            "early_stopping": True
        }
    }
}

yaml_path = os.path.join(OUTPUT_DIR, PREFIX + "config.yaml")

with open(yaml_path, "w") as f:
    yaml.dump(yaml_config, f, sort_keys=False)

# ============================================================
# GRAPHS
# ============================================================

optimization_graph_path = os.path.join(
    OUTPUT_DIR,
    PREFIX + "optimization_progress_graph.png"
)

plt.figure(figsize=(9, 5))
plt.plot(bis_progress, marker="o")
plt.title("BA + AIS Optimization Progress Graph")
plt.xlabel("Generation")
plt.ylabel("Best R2 Score")
plt.tight_layout()
plt.savefig(optimization_graph_path, dpi=300)
plt.show()

accuracy_graph_path = os.path.join(OUTPUT_DIR, PREFIX + "accuracy_graph.png")

plt.figure(figsize=(8, 5))
sns.barplot(x="Model", y="Accuracy_Percentage", data=results_df)
plt.title("BA + AIS Model Accuracy Comparison")
plt.xlabel("Model")
plt.ylabel("Accuracy Percentage")
plt.xticks(rotation=15)
plt.tight_layout()
plt.savefig(accuracy_graph_path, dpi=300)
plt.show()

result_graph_path = os.path.join(OUTPUT_DIR, PREFIX + "result_graph.png")

plt.figure(figsize=(8, 5))
sns.barplot(x="Model", y="RMSE", data=results_df)
plt.title("BA + AIS Model RMSE Result Comparison")
plt.xlabel("Model")
plt.ylabel("RMSE")
plt.xticks(rotation=15)
plt.tight_layout()
plt.savefig(result_graph_path, dpi=300)
plt.show()

comparison_graph_path = os.path.join(OUTPUT_DIR, PREFIX + "comparison_graph.png")

plt.figure(figsize=(10, 6))
plt.plot(y_test.values, label="Actual", marker="o")
plt.plot(rf_pred, label="BA + AIS RF Predicted", marker="x")
plt.plot(mlp_pred, label="BA + AIS MLP Predicted", marker="s")
plt.title("BA + AIS Actual vs Predicted Comparison")
plt.xlabel("Test Sample Index")
plt.ylabel(target_col)
plt.legend()
plt.tight_layout()
plt.savefig(comparison_graph_path, dpi=300)
plt.show()

prediction_graph_path = os.path.join(OUTPUT_DIR, PREFIX + "prediction_graph.png")

plt.figure(figsize=(10, 6))
plt.scatter(y_test.values, rf_pred, label="BA + AIS Random Forest")
plt.scatter(y_test.values, mlp_pred, label="BA + AIS MLP")
plt.plot(
    [y_test.min(), y_test.max()],
    [y_test.min(), y_test.max()],
    linestyle="--",
    label="Perfect Prediction Line"
)
plt.title("BA + AIS Prediction Graph")
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.legend()
plt.tight_layout()
plt.savefig(prediction_graph_path, dpi=300)
plt.show()

heatmap_path = os.path.join(OUTPUT_DIR, PREFIX + "heatmap.png")

encoded_df = df.copy()

for col in encoded_df.columns:
    if encoded_df[col].dtype == "object":
        le = LabelEncoder()
        encoded_df[col] = le.fit_transform(encoded_df[col].astype(str))

plt.figure(figsize=(12, 8))
sns.heatmap(encoded_df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("BA + AIS Feature Correlation Heatmap")
plt.tight_layout()
plt.savefig(heatmap_path, dpi=300)
plt.show()

loss_graph_path = os.path.join(OUTPUT_DIR, PREFIX + "training_loss_graph.png")

plt.figure(figsize=(8, 5))
plt.plot(mlp_model.loss_curve_, label="MLP Training Loss")
plt.title("BA + AIS MLP Training Loss Graph")
plt.xlabel("Iteration")
plt.ylabel("Loss")
plt.legend()
plt.tight_layout()
plt.savefig(loss_graph_path, dpi=300)
plt.show()

# ============================================================
# FINAL OUTPUT
# ============================================================

print("\nAll BA + AIS Hybrid files saved successfully!")

print("RF PKL Model:", rf_pkl_path)
print("MLP PKL Model:", mlp_pkl_path)
print("H5 File:", h5_path)
print("Scaler:", scaler_path)
print("Encoders:", encoder_path)
print("Selected Features:", selected_features_path)
print("YAML Config:", yaml_path)
print("JSON Report:", json_path)
print("Result CSV:", result_csv_path)
print("Prediction CSV:", prediction_csv_path)
print("Optimization Graph:", optimization_graph_path)
print("Accuracy Graph:", accuracy_graph_path)
print("Result Graph:", result_graph_path)
print("Comparison Graph:", comparison_graph_path)
print("Prediction Graph:", prediction_graph_path)
print("Heatmap:", heatmap_path)
print("Training Loss Graph:", loss_graph_path)
```
