```python
import os
import json
import yaml
import joblib
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping

warnings.filterwarnings("ignore")

# ============================================================
# PATHS
# ============================================================

DATASET_PATH = r"C:\Users\NXTWAVE\Downloads\Chemical Production Forecasting\Production_of_Major_Chemicals_2024-25_1_0.csv"
OUTPUT_DIR = r"C:\Users\NXTWAVE\Downloads\Chemical Production Forecasting"

os.makedirs(OUTPUT_DIR, exist_ok=True)

PREFIX = "csa_"

# ============================================================
# LOAD DATA
# ============================================================

df = pd.read_csv(DATASET_PATH)
df.columns = df.columns.str.strip()

print("Dataset Loaded Successfully")
print("Shape:", df.shape)
print(df.head())

# ============================================================
# CLEAN DATA
# ============================================================

df = df.replace(["-", "NA", "N/A", "na", "n/a", "", " "], np.nan)

for col in df.columns:
    if df[col].dtype == "object":
        cleaned = (
            df[col]
            .astype(str)
            .str.replace(",", "", regex=False)
            .str.strip()
        )
        df[col] = pd.to_numeric(cleaned, errors="ignore")

for col in df.columns:
    if df[col].dtype == "object":
        df[col] = df[col].fillna("Unknown")
    else:
        df[col] = df[col].fillna(df[col].median())

# ============================================================
# TARGET SELECTION
# ============================================================

numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()

if len(numeric_cols) == 0:
    raise ValueError("No numeric column found for prediction.")

target_col = numeric_cols[-1]

print("Target Column Selected:", target_col)

X = df.drop(columns=[target_col])
y = df[target_col]

# ============================================================
# ENCODING
# ============================================================

label_encoders = {}

for col in X.columns:
    if X[col].dtype == "object":
        le = LabelEncoder()
        X[col] = le.fit_transform(X[col].astype(str))
        label_encoders[col] = le

feature_names = list(X.columns)

# ============================================================
# SCALING
# ============================================================

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ============================================================
# TRAIN TEST SPLIT
# ============================================================

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.2,
    random_state=42
)

# ============================================================
# CSA FEATURE SELECTION
# Clonal Selection Algorithm
# ============================================================

def evaluate_solution(solution, X_train, X_test, y_train, y_test):
    selected_features = np.where(solution == 1)[0]

    if len(selected_features) == 0:
        return -9999

    model = RandomForestRegressor(
        n_estimators=100,
        max_depth=10,
        random_state=42
    )

    model.fit(X_train[:, selected_features], y_train)
    pred = model.predict(X_test[:, selected_features])

    return r2_score(y_test, pred)


def csa_feature_selection(
    X_train,
    X_test,
    y_train,
    y_test,
    num_features,
    population_size=20,
    generations=30,
    clone_factor=5,
    mutation_rate=0.15,
    replace_rate=0.25
):
    population = np.random.randint(0, 2, size=(population_size, num_features))

    for i in range(population_size):
        if np.sum(population[i]) == 0:
            population[i, np.random.randint(0, num_features)] = 1

    best_solution = None
    best_score = -9999
    progress = []

    for generation in range(generations):
        scores = np.array([
            evaluate_solution(individual, X_train, X_test, y_train, y_test)
            for individual in population
        ])

        sorted_indices = np.argsort(scores)[::-1]
        population = population[sorted_indices]
        scores = scores[sorted_indices]

        if scores[0] > best_score:
            best_score = scores[0]
            best_solution = population[0].copy()

        progress.append(best_score)

        elite_count = max(2, population_size // 4)
        elites = population[:elite_count]

        clone_pool = []

        for rank, elite in enumerate(elites):
            clones_per_elite = clone_factor * (elite_count - rank)

            for _ in range(clones_per_elite):
                clone = elite.copy()

                adaptive_mutation_rate = mutation_rate * (rank + 1)

                mutation_mask = np.random.rand(num_features) < adaptive_mutation_rate
                clone[mutation_mask] = 1 - clone[mutation_mask]

                if np.sum(clone) == 0:
                    clone[np.random.randint(0, num_features)] = 1

                clone_pool.append(clone)

        if len(clone_pool) > 0:
            clone_scores = np.array([
                evaluate_solution(clone, X_train, X_test, y_train, y_test)
                for clone in clone_pool
            ])

            best_clone_indices = np.argsort(clone_scores)[::-1]
            best_clones = np.array(clone_pool)[best_clone_indices[:population_size - elite_count]]
        else:
            best_clones = np.random.randint(
                0,
                2,
                size=(population_size - elite_count, num_features)
            )

        new_population = np.vstack([elites, best_clones])

        replace_count = int(population_size * replace_rate)

        if replace_count > 0:
            random_new = np.random.randint(0, 2, size=(replace_count, num_features))

            for i in range(replace_count):
                if np.sum(random_new[i]) == 0:
                    random_new[i, np.random.randint(0, num_features)] = 1

            new_population[-replace_count:] = random_new

        population = new_population[:population_size]

        print(
            f"Generation {generation + 1}/{generations} | "
            f"Best R2 Score: {best_score:.4f} | "
            f"Selected Features: {np.sum(best_solution)}"
        )

    selected_indices = np.where(best_solution == 1)[0]

    return selected_indices, best_score, progress


selected_indices, csa_best_score, csa_progress = csa_feature_selection(
    X_train,
    X_test,
    y_train,
    y_test,
    num_features=X_train.shape[1],
    population_size=20,
    generations=30,
    clone_factor=5,
    mutation_rate=0.15,
    replace_rate=0.25
)

selected_features = [feature_names[i] for i in selected_indices]

print("\nCSA Selected Features:")
print(selected_features)

# ============================================================
# FILTER DATA USING CSA FEATURES
# ============================================================

X_train_csa = X_train[:, selected_indices]
X_test_csa = X_test[:, selected_indices]

# ============================================================
# RANDOM FOREST WITH CSA FEATURES
# ============================================================

rf_model = RandomForestRegressor(
    n_estimators=200,
    max_depth=10,
    random_state=42
)

rf_model.fit(X_train_csa, y_train)
rf_pred = rf_model.predict(X_test_csa)

rf_mae = mean_absolute_error(y_test, rf_pred)
rf_mse = mean_squared_error(y_test, rf_pred)
rf_rmse = np.sqrt(rf_mse)
rf_r2 = r2_score(y_test, rf_pred)

# ============================================================
# DEEP LEARNING WITH CSA FEATURES
# ============================================================

dl_model = Sequential([
    Dense(128, activation="relu", input_shape=(X_train_csa.shape[1],)),
    Dropout(0.2),
    Dense(64, activation="relu"),
    Dropout(0.2),
    Dense(32, activation="relu"),
    Dense(1)
])

dl_model.compile(
    optimizer="adam",
    loss="mse",
    metrics=["mae"]
)

early_stop = EarlyStopping(
    monitor="val_loss",
    patience=20,
    restore_best_weights=True
)

history = dl_model.fit(
    X_train_csa,
    y_train,
    validation_split=0.2,
    epochs=200,
    batch_size=8,
    callbacks=[early_stop],
    verbose=1
)

dl_pred = dl_model.predict(X_test_csa).flatten()

dl_mae = mean_absolute_error(y_test, dl_pred)
dl_mse = mean_squared_error(y_test, dl_pred)
dl_rmse = np.sqrt(dl_mse)
dl_r2 = r2_score(y_test, dl_pred)

# ============================================================
# SAVE MODELS
# ============================================================

h5_path = os.path.join(OUTPUT_DIR, PREFIX + "model.h5")
pkl_path = os.path.join(OUTPUT_DIR, PREFIX + "random_forest_model.pkl")
scaler_path = os.path.join(OUTPUT_DIR, PREFIX + "scaler.pkl")
encoder_path = os.path.join(OUTPUT_DIR, PREFIX + "label_encoders.pkl")
selected_features_path = os.path.join(OUTPUT_DIR, PREFIX + "selected_features.pkl")

dl_model.save(h5_path)
joblib.dump(rf_model, pkl_path)
joblib.dump(scaler, scaler_path)
joblib.dump(label_encoders, encoder_path)
joblib.dump(selected_features, selected_features_path)

# ============================================================
# PREDICTION CSV
# ============================================================

prediction_df = pd.DataFrame({
    "Actual": y_test.values,
    "CSA_Random_Forest_Predicted": rf_pred,
    "CSA_Deep_Learning_Predicted": dl_pred
})

prediction_csv_path = os.path.join(OUTPUT_DIR, PREFIX + "prediction.csv")
prediction_df.to_csv(prediction_csv_path, index=False)

# ============================================================
# RESULT CSV
# ============================================================

results_df = pd.DataFrame({
    "Model": ["CSA + Random Forest", "CSA + Deep Learning H5"],
    "MAE": [rf_mae, dl_mae],
    "MSE": [rf_mse, dl_mse],
    "RMSE": [rf_rmse, dl_rmse],
    "R2_Score": [rf_r2, dl_r2],
    "Accuracy_Percentage": [rf_r2 * 100, dl_r2 * 100],
    "Selected_Features_Count": [len(selected_features), len(selected_features)]
})

result_csv_path = os.path.join(OUTPUT_DIR, PREFIX + "result.csv")
results_df.to_csv(result_csv_path, index=False)

print("\nCSA Result CSV:")
print(results_df)

# ============================================================
# JSON REPORT
# ============================================================

json_report = {
    "project_name": "Chemical Production Forecasting System using CSA",
    "author": "Sagnik Patra",
    "algorithm": "Clonal Selection Algorithm Feature Selection",
    "target_column": target_col,
    "rows": int(df.shape[0]),
    "columns": int(df.shape[1]),
    "total_features": int(len(feature_names)),
    "selected_features_count": int(len(selected_features)),
    "selected_features": selected_features,
    "csa_best_score": float(csa_best_score),
    "random_forest": {
        "mae": float(rf_mae),
        "mse": float(rf_mse),
        "rmse": float(rf_rmse),
        "r2_score": float(rf_r2),
        "accuracy_percentage": float(rf_r2 * 100)
    },
    "deep_learning_h5": {
        "mae": float(dl_mae),
        "mse": float(dl_mse),
        "rmse": float(dl_rmse),
        "r2_score": float(dl_r2),
        "accuracy_percentage": float(dl_r2 * 100)
    }
}

json_path = os.path.join(OUTPUT_DIR, PREFIX + "report.json")

with open(json_path, "w") as f:
    json.dump(json_report, f, indent=4)

# ============================================================
# YAML CONFIG
# ============================================================

yaml_config = {
    "project": {
        "name": "Chemical Production Forecasting System using CSA",
        "author": "Sagnik Patra"
    },
    "paths": {
        "dataset_path": DATASET_PATH,
        "output_directory": OUTPUT_DIR,
        "h5_model": h5_path,
        "pkl_model": pkl_path,
        "prediction_csv": prediction_csv_path,
        "result_csv": result_csv_path,
        "json_report": json_path
    },
    "data": {
        "target_column": target_col,
        "rows": int(df.shape[0]),
        "columns": int(df.shape[1]),
        "all_features": feature_names,
        "selected_features": selected_features
    },
    "csa_parameters": {
        "population_size": 20,
        "generations": 30,
        "clone_factor": 5,
        "mutation_rate": 0.15,
        "replace_rate": 0.25
    },
    "models": {
        "random_forest": {
            "n_estimators": 200,
            "max_depth": 10,
            "random_state": 42
        },
        "deep_learning": {
            "layers": [128, 64, 32, 1],
            "activation": "relu",
            "optimizer": "adam",
            "loss": "mse",
            "epochs": 200,
            "batch_size": 8
        }
    }
}

yaml_path = os.path.join(OUTPUT_DIR, PREFIX + "config.yaml")

with open(yaml_path, "w") as f:
    yaml.dump(yaml_config, f, sort_keys=False)

# ============================================================
# GRAPH 1: CSA OPTIMIZATION PROGRESS GRAPH
# ============================================================

csa_progress_graph_path = os.path.join(OUTPUT_DIR, PREFIX + "optimization_progress_graph.png")

plt.figure(figsize=(9, 5))
plt.plot(csa_progress, marker="o")
plt.title("CSA Optimization Progress Graph")
plt.xlabel("Generation")
plt.ylabel("Best R2 Score")
plt.tight_layout()
plt.savefig(csa_progress_graph_path, dpi=300)
plt.show()

# ============================================================
# GRAPH 2: ACCURACY GRAPH
# ============================================================

accuracy_graph_path = os.path.join(OUTPUT_DIR, PREFIX + "accuracy_graph.png")

plt.figure(figsize=(8, 5))
sns.barplot(
    x="Model",
    y="Accuracy_Percentage",
    data=results_df
)
plt.title("CSA Model Accuracy Comparison")
plt.xlabel("Model")
plt.ylabel("Accuracy Percentage")
plt.xticks(rotation=15)
plt.tight_layout()
plt.savefig(accuracy_graph_path, dpi=300)
plt.show()

# ============================================================
# GRAPH 3: RESULT GRAPH - RMSE
# ============================================================

result_graph_path = os.path.join(OUTPUT_DIR, PREFIX + "result_graph.png")

plt.figure(figsize=(8, 5))
sns.barplot(
    x="Model",
    y="RMSE",
    data=results_df
)
plt.title("CSA Model RMSE Result Comparison")
plt.xlabel("Model")
plt.ylabel("RMSE")
plt.xticks(rotation=15)
plt.tight_layout()
plt.savefig(result_graph_path, dpi=300)
plt.show()

# ============================================================
# GRAPH 4: COMPARISON GRAPH
# ============================================================

comparison_graph_path = os.path.join(OUTPUT_DIR, PREFIX + "comparison_graph.png")

plt.figure(figsize=(10, 6))
plt.plot(y_test.values, label="Actual", marker="o")
plt.plot(rf_pred, label="CSA Random Forest Predicted", marker="x")
plt.plot(dl_pred, label="CSA Deep Learning Predicted", marker="s")
plt.title("CSA Actual vs Predicted Comparison")
plt.xlabel("Test Sample Index")
plt.ylabel(target_col)
plt.legend()
plt.tight_layout()
plt.savefig(comparison_graph_path, dpi=300)
plt.show()

# ============================================================
# GRAPH 5: PREDICTION GRAPH
# ============================================================

prediction_graph_path = os.path.join(OUTPUT_DIR, PREFIX + "prediction_graph.png")

plt.figure(figsize=(10, 6))
plt.scatter(y_test.values, rf_pred, label="CSA Random Forest")
plt.scatter(y_test.values, dl_pred, label="CSA Deep Learning")
plt.plot(
    [y_test.min(), y_test.max()],
    [y_test.min(), y_test.max()],
    linestyle="--",
    label="Perfect Prediction Line"
)
plt.title("CSA Prediction Graph")
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.legend()
plt.tight_layout()
plt.savefig(prediction_graph_path, dpi=300)
plt.show()

# ============================================================
# GRAPH 6: HEATMAP
# ============================================================

heatmap_path = os.path.join(OUTPUT_DIR, PREFIX + "heatmap.png")

encoded_df = df.copy()

for col in encoded_df.columns:
    if encoded_df[col].dtype == "object":
        le = LabelEncoder()
        encoded_df[col] = le.fit_transform(encoded_df[col].astype(str))

plt.figure(figsize=(12, 8))
sns.heatmap(
    encoded_df.corr(),
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)
plt.title("CSA Feature Correlation Heatmap")
plt.tight_layout()
plt.savefig(heatmap_path, dpi=300)
plt.show()

# ============================================================
# GRAPH 7: TRAINING LOSS GRAPH
# ============================================================

loss_graph_path = os.path.join(OUTPUT_DIR, PREFIX + "training_loss_graph.png")

plt.figure(figsize=(8, 5))
plt.plot(history.history["loss"], label="Training Loss")
plt.plot(history.history["val_loss"], label="Validation Loss")
plt.title("CSA Deep Learning Training Loss Graph")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.tight_layout()
plt.savefig(loss_graph_path, dpi=300)
plt.show()

# ============================================================
# FINAL OUTPUT
# ============================================================

print("\nAll CSA files saved successfully!")

print("\nSaved Files:")
print("H5 Model:", h5_path)
print("PKL Model:", pkl_path)
print("Scaler:", scaler_path)
print("Encoders:", encoder_path)
print("Selected Features:", selected_features_path)
print("YAML Config:", yaml_path)
print("JSON Report:", json_path)
print("Result CSV:", result_csv_path)
print("Prediction CSV:", prediction_csv_path)
print("CSA Progress Graph:", csa_progress_graph_path)
print("Accuracy Graph:", accuracy_graph_path)
print("Result Graph:", result_graph_path)
print("Comparison Graph:", comparison_graph_path)
print("Prediction Graph:", prediction_graph_path)
print("Heatmap:", heatmap_path)
print("Training Loss Graph:", loss_graph_path)
```
