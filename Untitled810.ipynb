```python
import os
import json
import yaml
import h5py
import joblib
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

warnings.filterwarnings("ignore")

DATASET_PATH = r"C:\Users\NXTWAVE\Downloads\Chemical Production Forecasting\Production_of_Major_Chemicals_2024-25_1_0.csv"
OUTPUT_DIR = r"C:\Users\NXTWAVE\Downloads\Chemical Production Forecasting"

os.makedirs(OUTPUT_DIR, exist_ok=True)

PREFIX = "wloa_"

# ============================================================
# LOAD DATA
# ============================================================

df = pd.read_csv(DATASET_PATH)
df.columns = df.columns.str.strip()

print("Dataset Loaded Successfully")
print("Shape:", df.shape)
print(df.head())

# ============================================================
# CLEAN DATA
# ============================================================

df = df.replace(["-", "NA", "N/A", "na", "n/a", "", " "], np.nan)

for col in df.columns:
    if df[col].dtype == "object":
        cleaned = df[col].astype(str).str.replace(",", "", regex=False).str.strip()
        df[col] = pd.to_numeric(cleaned, errors="ignore")

for col in df.columns:
    if df[col].dtype == "object":
        df[col] = df[col].fillna("Unknown")
    else:
        df[col] = df[col].fillna(df[col].median())

# ============================================================
# TARGET SELECTION
# ============================================================

numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()

if len(numeric_cols) == 0:
    raise ValueError("No numeric column found for prediction.")

target_col = numeric_cols[-1]
print("Target Column Selected:", target_col)

X = df.drop(columns=[target_col])
y = df[target_col]

# ============================================================
# ENCODING
# ============================================================

label_encoders = {}

for col in X.columns:
    if X[col].dtype == "object":
        le = LabelEncoder()
        X[col] = le.fit_transform(X[col].astype(str))
        label_encoders[col] = le

feature_names = list(X.columns)

# ============================================================
# SCALING
# ============================================================

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ============================================================
# TRAIN TEST SPLIT
# ============================================================

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.2,
    random_state=42
)

# ============================================================
# FITNESS FUNCTION
# ============================================================

def evaluate_solution(solution, X_train, X_test, y_train, y_test):
    selected_features = np.where(solution == 1)[0]

    if len(selected_features) == 0:
        return -9999

    model = RandomForestRegressor(
        n_estimators=100,
        max_depth=10,
        random_state=42
    )

    model.fit(X_train[:, selected_features], y_train)
    pred = model.predict(X_test[:, selected_features])

    return r2_score(y_test, pred)

# ============================================================
# HYBRID WOA + ALOA FEATURE SELECTION
# Whale Optimization Algorithm + Ant Lion Optimization Algorithm
# ============================================================

def sigmoid(x):
    x = np.clip(x, -500, 500)
    return 1 / (1 + np.exp(-x))


def roulette_wheel_selection(fitness):
    fitness = np.array(fitness)

    min_fit = np.min(fitness)
    if min_fit < 0:
        fitness = fitness - min_fit + 1e-6

    total = np.sum(fitness)

    if total == 0:
        return np.random.randint(0, len(fitness))

    probabilities = fitness / total
    return np.random.choice(len(fitness), p=probabilities)


def binary_random_walk(num_features, selected_antlion, elite, iteration, max_iter):
    ant = np.random.randint(0, 2, num_features)

    influence_rate = 0.3 + 0.7 * (iteration / max_iter)

    for j in range(num_features):
        r = np.random.rand()

        if r < influence_rate / 2:
            ant[j] = selected_antlion[j]
        elif r < influence_rate:
            ant[j] = elite[j]

    mutation_rate = 0.2 * (1 - iteration / max_iter)
    mutation_mask = np.random.rand(num_features) < mutation_rate
    ant[mutation_mask] = 1 - ant[mutation_mask]

    if np.sum(ant) == 0:
        ant[np.random.randint(0, num_features)] = 1

    return ant


def woa_aloa_feature_selection(
    X_train,
    X_test,
    y_train,
    y_test,
    num_features,
    population_size=30,
    generations=30
):
    whales = np.random.randint(0, 2, size=(population_size, num_features))

    for i in range(population_size):
        if np.sum(whales[i]) == 0:
            whales[i, np.random.randint(0, num_features)] = 1

    whale_scores = np.array([
        evaluate_solution(whale, X_train, X_test, y_train, y_test)
        for whale in whales
    ])

    best_index = np.argmax(whale_scores)
    best_whale = whales[best_index].copy()
    best_score = whale_scores[best_index]

    progress = []

    for generation in range(generations):

        a = 2 - generation * (2 / generations)

        new_whales = []

        # ====================================================
        # WOA PHASE
        # ====================================================

        for i in range(population_size):
            r1 = np.random.rand(num_features)
            r2 = np.random.rand(num_features)

            A = 2 * a * r1 - a
            C = 2 * r2

            p = np.random.rand()
            l = np.random.uniform(-1, 1)
            b = 1

            current = whales[i].copy()

            if p < 0.5:
                if np.mean(np.abs(A)) < 1:
                    D = np.abs(C * best_whale - current)
                    continuous_position = best_whale - A * D
                else:
                    random_index = np.random.randint(0, population_size)
                    random_whale = whales[random_index]
                    D = np.abs(C * random_whale - current)
                    continuous_position = random_whale - A * D
            else:
                D = np.abs(best_whale - current)
                continuous_position = (
                    D * np.exp(b * l) * np.cos(2 * np.pi * l) + best_whale
                )

            probabilities = sigmoid(continuous_position)
            new_solution = (
                np.random.rand(num_features) < probabilities
            ).astype(int)

            if np.sum(new_solution) == 0:
                new_solution[np.random.randint(0, num_features)] = 1

            new_whales.append(new_solution)

        new_whales = np.array(new_whales)

        # ====================================================
        # ALOA PHASE
        # ====================================================

        antlion_scores = np.array([
            evaluate_solution(whale, X_train, X_test, y_train, y_test)
            for whale in new_whales
        ])

        ants = []

        for _ in range(population_size):
            selected_index = roulette_wheel_selection(antlion_scores)
            selected_antlion = new_whales[selected_index]

            ant = binary_random_walk(
                num_features,
                selected_antlion,
                best_whale,
                generation,
                generations
            )

            ants.append(ant)

        ants = np.array(ants)

        ant_scores = np.array([
            evaluate_solution(ant, X_train, X_test, y_train, y_test)
            for ant in ants
        ])

        # ====================================================
        # SELECTION PHASE
        # ====================================================

        combined_population = np.vstack([whales, new_whales, ants])
        combined_scores = np.concatenate([whale_scores, antlion_scores, ant_scores])

        sorted_indices = np.argsort(combined_scores)[::-1]

        whales = combined_population[sorted_indices[:population_size]]
        whale_scores = combined_scores[sorted_indices[:population_size]]

        if whale_scores[0] > best_score:
            best_score = whale_scores[0]
            best_whale = whales[0].copy()

        progress.append(best_score)

        print(
            f"Generation {generation + 1}/{generations} | "
            f"Best R2 Score: {best_score:.4f} | "
            f"Selected Features: {np.sum(best_whale)}"
        )

    selected_indices = np.where(best_whale == 1)[0]

    return selected_indices, best_score, progress


selected_indices, wloa_best_score, wloa_progress = woa_aloa_feature_selection(
    X_train,
    X_test,
    y_train,
    y_test,
    num_features=X_train.shape[1],
    population_size=30,
    generations=30
)

selected_features = [feature_names[i] for i in selected_indices]

print("\nWLOA Selected Features:")
print(selected_features)

X_train_wloa = X_train[:, selected_indices]
X_test_wloa = X_test[:, selected_indices]

# ============================================================
# MODEL 1: RANDOM FOREST
# ============================================================

rf_model = RandomForestRegressor(
    n_estimators=200,
    max_depth=10,
    random_state=42
)

rf_model.fit(X_train_wloa, y_train)
rf_pred = rf_model.predict(X_test_wloa)

rf_mae = mean_absolute_error(y_test, rf_pred)
rf_mse = mean_squared_error(y_test, rf_pred)
rf_rmse = np.sqrt(rf_mse)
rf_r2 = r2_score(y_test, rf_pred)

# ============================================================
# MODEL 2: MLP REGRESSOR
# ============================================================

mlp_model = MLPRegressor(
    hidden_layer_sizes=(128, 64, 32),
    activation="relu",
    solver="adam",
    max_iter=1000,
    random_state=42,
    early_stopping=True
)

mlp_model.fit(X_train_wloa, y_train)
mlp_pred = mlp_model.predict(X_test_wloa)

mlp_mae = mean_absolute_error(y_test, mlp_pred)
mlp_mse = mean_squared_error(y_test, mlp_pred)
mlp_rmse = np.sqrt(mlp_mse)
mlp_r2 = r2_score(y_test, mlp_pred)

# ============================================================
# SAVE MODELS
# ============================================================

rf_pkl_path = os.path.join(OUTPUT_DIR, PREFIX + "random_forest_model.pkl")
mlp_pkl_path = os.path.join(OUTPUT_DIR, PREFIX + "mlp_model.pkl")
scaler_path = os.path.join(OUTPUT_DIR, PREFIX + "scaler.pkl")
encoder_path = os.path.join(OUTPUT_DIR, PREFIX + "label_encoders.pkl")
selected_features_path = os.path.join(OUTPUT_DIR, PREFIX + "selected_features.pkl")
h5_path = os.path.join(OUTPUT_DIR, PREFIX + "model_data.h5")

joblib.dump(rf_model, rf_pkl_path)
joblib.dump(mlp_model, mlp_pkl_path)
joblib.dump(scaler, scaler_path)
joblib.dump(label_encoders, encoder_path)
joblib.dump(selected_features, selected_features_path)

with h5py.File(h5_path, "w") as h5:
    h5.create_dataset("selected_feature_indices", data=selected_indices)
    h5.create_dataset("rf_predictions", data=rf_pred)
    h5.create_dataset("mlp_predictions", data=mlp_pred)
    h5.create_dataset("actual_values", data=y_test.values)
    h5.attrs["project_name"] = "Chemical Production Forecasting using WOA + ALOA"
    h5.attrs["target_column"] = target_col
    h5.attrs["wloa_best_score"] = wloa_best_score

# ============================================================
# SAVE CSV FILES
# ============================================================

prediction_df = pd.DataFrame({
    "Actual": y_test.values,
    "WLOA_Random_Forest_Predicted": rf_pred,
    "WLOA_MLP_Predicted": mlp_pred
})

prediction_csv_path = os.path.join(OUTPUT_DIR, PREFIX + "prediction.csv")
prediction_df.to_csv(prediction_csv_path, index=False)

results_df = pd.DataFrame({
    "Model": ["WOA + ALOA + Random Forest", "WOA + ALOA + MLP Regressor"],
    "MAE": [rf_mae, mlp_mae],
    "MSE": [rf_mse, mlp_mse],
    "RMSE": [rf_rmse, mlp_rmse],
    "R2_Score": [rf_r2, mlp_r2],
    "Accuracy_Percentage": [rf_r2 * 100, mlp_r2 * 100],
    "Selected_Features_Count": [len(selected_features), len(selected_features)]
})

result_csv_path = os.path.join(OUTPUT_DIR, PREFIX + "result.csv")
results_df.to_csv(result_csv_path, index=False)

print("\nWLOA Result CSV:")
print(results_df)

# ============================================================
# JSON REPORT
# ============================================================

json_report = {
    "project_name": "Chemical Production Forecasting System using WOA + ALOA",
    "author": "Sagnik Patra",
    "algorithm": "Hybrid Whale Optimization Algorithm + Ant Lion Optimization Algorithm",
    "prefix": PREFIX,
    "target_column": target_col,
    "rows": int(df.shape[0]),
    "columns": int(df.shape[1]),
    "total_features": int(len(feature_names)),
    "selected_features_count": int(len(selected_features)),
    "selected_features": selected_features,
    "wloa_best_score": float(wloa_best_score),
    "random_forest": {
        "mae": float(rf_mae),
        "mse": float(rf_mse),
        "rmse": float(rf_rmse),
        "r2_score": float(rf_r2),
        "accuracy_percentage": float(rf_r2 * 100)
    },
    "mlp_regressor": {
        "mae": float(mlp_mae),
        "mse": float(mlp_mse),
        "rmse": float(mlp_rmse),
        "r2_score": float(mlp_r2),
        "accuracy_percentage": float(mlp_r2 * 100)
    }
}

json_path = os.path.join(OUTPUT_DIR, PREFIX + "report.json")

with open(json_path, "w") as f:
    json.dump(json_report, f, indent=4)

# ============================================================
# YAML CONFIG
# ============================================================

yaml_config = {
    "project": {
        "name": "Chemical Production Forecasting System using WOA + ALOA",
        "author": "Sagnik Patra"
    },
    "prefix": PREFIX,
    "paths": {
        "dataset_path": DATASET_PATH,
        "output_directory": OUTPUT_DIR,
        "h5_file": h5_path,
        "rf_pkl_model": rf_pkl_path,
        "mlp_pkl_model": mlp_pkl_path,
        "prediction_csv": prediction_csv_path,
        "result_csv": result_csv_path,
        "json_report": json_path
    },
    "data": {
        "target_column": target_col,
        "rows": int(df.shape[0]),
        "columns": int(df.shape[1]),
        "all_features": feature_names,
        "selected_features": selected_features
    },
    "hybrid_parameters": {
        "algorithm": "WOA + ALOA",
        "population_size": 30,
        "generations": 30
    },
    "models": {
        "random_forest": {
            "n_estimators": 200,
            "max_depth": 10,
            "random_state": 42
        },
        "mlp_regressor": {
            "hidden_layer_sizes": [128, 64, 32],
            "activation": "relu",
            "solver": "adam",
            "max_iter": 1000,
            "early_stopping": True
        }
    }
}

yaml_path = os.path.join(OUTPUT_DIR, PREFIX + "config.yaml")

with open(yaml_path, "w") as f:
    yaml.dump(yaml_config, f, sort_keys=False)

# ============================================================
# GRAPHS
# ============================================================

optimization_graph_path = os.path.join(
    OUTPUT_DIR,
    PREFIX + "optimization_progress_graph.png"
)

plt.figure(figsize=(9, 5))
plt.plot(wloa_progress, marker="o")
plt.title("WOA + ALOA Optimization Progress Graph")
plt.xlabel("Generation")
plt.ylabel("Best R2 Score")
plt.tight_layout()
plt.savefig(optimization_graph_path, dpi=300)
plt.show()

accuracy_graph_path = os.path.join(OUTPUT_DIR, PREFIX + "accuracy_graph.png")

plt.figure(figsize=(8, 5))
sns.barplot(x="Model", y="Accuracy_Percentage", data=results_df)
plt.title("WOA + ALOA Model Accuracy Comparison")
plt.xlabel("Model")
plt.ylabel("Accuracy Percentage")
plt.xticks(rotation=15)
plt.tight_layout()
plt.savefig(accuracy_graph_path, dpi=300)
plt.show()

result_graph_path = os.path.join(OUTPUT_DIR, PREFIX + "result_graph.png")

plt.figure(figsize=(8, 5))
sns.barplot(x="Model", y="RMSE", data=results_df)
plt.title("WOA + ALOA Model RMSE Result Comparison")
plt.xlabel("Model")
plt.ylabel("RMSE")
plt.xticks(rotation=15)
plt.tight_layout()
plt.savefig(result_graph_path, dpi=300)
plt.show()

comparison_graph_path = os.path.join(OUTPUT_DIR, PREFIX + "comparison_graph.png")

plt.figure(figsize=(10, 6))
plt.plot(y_test.values, label="Actual", marker="o")
plt.plot(rf_pred, label="WOA + ALOA RF Predicted", marker="x")
plt.plot(mlp_pred, label="WOA + ALOA MLP Predicted", marker="s")
plt.title("WOA + ALOA Actual vs Predicted Comparison")
plt.xlabel("Test Sample Index")
plt.ylabel(target_col)
plt.legend()
plt.tight_layout()
plt.savefig(comparison_graph_path, dpi=300)
plt.show()

prediction_graph_path = os.path.join(OUTPUT_DIR, PREFIX + "prediction_graph.png")

plt.figure(figsize=(10, 6))
plt.scatter(y_test.values, rf_pred, label="WOA + ALOA Random Forest")
plt.scatter(y_test.values, mlp_pred, label="WOA + ALOA MLP")
plt.plot(
    [y_test.min(), y_test.max()],
    [y_test.min(), y_test.max()],
    linestyle="--",
    label="Perfect Prediction Line"
)
plt.title("WOA + ALOA Prediction Graph")
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.legend()
plt.tight_layout()
plt.savefig(prediction_graph_path, dpi=300)
plt.show()

heatmap_path = os.path.join(OUTPUT_DIR, PREFIX + "heatmap.png")

encoded_df = df.copy()

for col in encoded_df.columns:
    if encoded_df[col].dtype == "object":
        le = LabelEncoder()
        encoded_df[col] = le.fit_transform(encoded_df[col].astype(str))

plt.figure(figsize=(12, 8))
sns.heatmap(encoded_df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("WOA + ALOA Feature Correlation Heatmap")
plt.tight_layout()
plt.savefig(heatmap_path, dpi=300)
plt.show()

loss_graph_path = os.path.join(OUTPUT_DIR, PREFIX + "training_loss_graph.png")

plt.figure(figsize=(8, 5))
plt.plot(mlp_model.loss_curve_, label="MLP Training Loss")
plt.title("WOA + ALOA MLP Training Loss Graph")
plt.xlabel("Iteration")
plt.ylabel("Loss")
plt.legend()
plt.tight_layout()
plt.savefig(loss_graph_path, dpi=300)
plt.show()

# ============================================================
# FINAL OUTPUT
# ============================================================

print("\nAll WOA + ALOA Hybrid files saved successfully!")

print("RF PKL Model:", rf_pkl_path)
print("MLP PKL Model:", mlp_pkl_path)
print("H5 File:", h5_path)
print("Scaler:", scaler_path)
print("Encoders:", encoder_path)
print("Selected Features:", selected_features_path)
print("YAML Config:", yaml_path)
print("JSON Report:", json_path)
print("Result CSV:", result_csv_path)
print("Prediction CSV:", prediction_csv_path)
print("Optimization Graph:", optimization_graph_path)
print("Accuracy Graph:", accuracy_graph_path)
print("Result Graph:", result_graph_path)
print("Comparison Graph:", comparison_graph_path)
print("Prediction Graph:", prediction_graph_path)
print("Heatmap:", heatmap_path)
print("Training Loss Graph:", loss_graph_path)
```
