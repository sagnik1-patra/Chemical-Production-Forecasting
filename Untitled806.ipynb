```python
import os
import json
import yaml
import h5py
import joblib
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

warnings.filterwarnings("ignore")

# ============================================================
# PATHS
# ============================================================

DATASET_PATH = r"C:\Users\NXTWAVE\Downloads\Chemical Production Forecasting\Production_of_Major_Chemicals_2024-25_1_0.csv"
OUTPUT_DIR = r"C:\Users\NXTWAVE\Downloads\Chemical Production Forecasting"

os.makedirs(OUTPUT_DIR, exist_ok=True)

PREFIX = "pis_"

# ============================================================
# LOAD DATA
# ============================================================

df = pd.read_csv(DATASET_PATH)
df.columns = df.columns.str.strip()

print("Dataset Loaded Successfully")
print("Shape:", df.shape)

# ============================================================
# CLEAN DATA
# ============================================================

df = df.replace(["-", "NA", "N/A", "na", "n/a", "", " "], np.nan)

for col in df.columns:
    if df[col].dtype == "object":
        cleaned = (
            df[col]
            .astype(str)
            .str.replace(",", "", regex=False)
            .str.strip()
        )
        df[col] = pd.to_numeric(cleaned, errors="ignore")

for col in df.columns:
    if df[col].dtype == "object":
        df[col] = df[col].fillna("Unknown")
    else:
        df[col] = df[col].fillna(df[col].median())

# ============================================================
# TARGET COLUMN
# ============================================================

numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()

if len(numeric_cols) == 0:
    raise ValueError("No numeric target column found.")

target_col = numeric_cols[-1]

print("Target Column:", target_col)

X = df.drop(columns=[target_col])
y = df[target_col]

# ============================================================
# ENCODING
# ============================================================

label_encoders = {}

for col in X.columns:
    if X[col].dtype == "object":
        le = LabelEncoder()
        X[col] = le.fit_transform(X[col].astype(str))
        label_encoders[col] = le

feature_names = list(X.columns)

# ============================================================
# SCALING
# ============================================================

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ============================================================
# SPLIT
# ============================================================

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.2,
    random_state=42
)

# ============================================================
# FITNESS FUNCTION
# ============================================================

def evaluate_solution(solution, X_train, X_test, y_train, y_test):
    selected = np.where(solution == 1)[0]

    if len(selected) == 0:
        return -9999

    model = RandomForestRegressor(
        n_estimators=100,
        max_depth=10,
        random_state=42
    )

    model.fit(X_train[:, selected], y_train)

    pred = model.predict(X_test[:, selected])

    return r2_score(y_test, pred)

# ============================================================
# AIS + PSO HYBRID FEATURE SELECTION
# ============================================================

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def ais_pso_feature_selection(
    X_train,
    X_test,
    y_train,
    y_test,
    num_features,
    population_size=25,
    generations=30,
    inertia_weight=0.7,
    cognitive_coeff=1.5,
    social_coeff=1.5,
    clone_factor=4,
    mutation_rate=0.1
):

    population = np.random.randint(
        0,
        2,
        size=(population_size, num_features)
    )

    velocities = np.random.uniform(
        -1,
        1,
        size=(population_size, num_features)
    )

    for i in range(population_size):
        if np.sum(population[i]) == 0:
            population[i, np.random.randint(0, num_features)] = 1

    personal_best = population.copy()

    personal_best_scores = np.array([
        evaluate_solution(ind, X_train, X_test, y_train, y_test)
        for ind in population
    ])

    best_idx = np.argmax(personal_best_scores)

    global_best = personal_best[best_idx].copy()
    global_best_score = personal_best_scores[best_idx]

    progress = []

    for generation in range(generations):

        # ============================
        # PSO UPDATE
        # ============================

        for i in range(population_size):

            r1 = np.random.rand(num_features)
            r2 = np.random.rand(num_features)

            velocities[i] = (
                inertia_weight * velocities[i]
                + cognitive_coeff * r1 * (personal_best[i] - population[i])
                + social_coeff * r2 * (global_best - population[i])
            )

            probabilities = sigmoid(velocities[i])

            population[i] = (
                np.random.rand(num_features) < probabilities
            ).astype(int)

            if np.sum(population[i]) == 0:
                population[i, np.random.randint(0, num_features)] = 1

        # ============================
        # AIS CLONING + MUTATION
        # ============================

        clones = []

        scores = np.array([
            evaluate_solution(ind, X_train, X_test, y_train, y_test)
            for ind in population
        ])

        sorted_indices = np.argsort(scores)[::-1]

        elites = population[sorted_indices[:5]]

        for elite in elites:
            for _ in range(clone_factor):

                clone = elite.copy()

                mutation_mask = (
                    np.random.rand(num_features) < mutation_rate
                )

                clone[mutation_mask] = 1 - clone[mutation_mask]

                if np.sum(clone) == 0:
                    clone[np.random.randint(0, num_features)] = 1

                clones.append(clone)

        clones = np.array(clones)

        clone_scores = np.array([
            evaluate_solution(ind, X_train, X_test, y_train, y_test)
            for ind in clones
        ])

        best_clone_idx = np.argmax(clone_scores)

        if clone_scores[best_clone_idx] > global_best_score:
            global_best_score = clone_scores[best_clone_idx]
            global_best = clones[best_clone_idx].copy()

        # ============================
        # UPDATE PERSONAL BEST
        # ============================

        for i in range(population_size):

            score = evaluate_solution(
                population[i],
                X_train,
                X_test,
                y_train,
                y_test
            )

            if score > personal_best_scores[i]:
                personal_best_scores[i] = score
                personal_best[i] = population[i].copy()

                if score > global_best_score:
                    global_best_score = score
                    global_best = population[i].copy()

        progress.append(global_best_score)

        print(
            f"Generation {generation+1}/{generations} | "
            f"Best R2 Score: {global_best_score:.4f} | "
            f"Selected Features: {np.sum(global_best)}"
        )

    selected_indices = np.where(global_best == 1)[0]

    return selected_indices, global_best_score, progress

# ============================================================
# RUN HYBRID AIS + PSO
# ============================================================

selected_indices, pis_best_score, pis_progress = ais_pso_feature_selection(
    X_train,
    X_test,
    y_train,
    y_test,
    num_features=X_train.shape[1]
)

selected_features = [feature_names[i] for i in selected_indices]

print("\nSelected Features:")
print(selected_features)

# ============================================================
# FILTER FEATURES
# ============================================================

X_train_pis = X_train[:, selected_indices]
X_test_pis = X_test[:, selected_indices]

# ============================================================
# RANDOM FOREST MODEL
# ============================================================

rf_model = RandomForestRegressor(
    n_estimators=200,
    max_depth=10,
    random_state=42
)

rf_model.fit(X_train_pis, y_train)

rf_pred = rf_model.predict(X_test_pis)

rf_mae = mean_absolute_error(y_test, rf_pred)
rf_mse = mean_squared_error(y_test, rf_pred)
rf_rmse = np.sqrt(rf_mse)
rf_r2 = r2_score(y_test, rf_pred)

# ============================================================
# MLP MODEL
# ============================================================

mlp_model = MLPRegressor(
    hidden_layer_sizes=(128, 64, 32),
    activation="relu",
    solver="adam",
    max_iter=1000,
    random_state=42,
    early_stopping=True
)

mlp_model.fit(X_train_pis, y_train)

mlp_pred = mlp_model.predict(X_test_pis)

mlp_mae = mean_absolute_error(y_test, mlp_pred)
mlp_mse = mean_squared_error(y_test, mlp_pred)
mlp_rmse = np.sqrt(mlp_mse)
mlp_r2 = r2_score(y_test, mlp_pred)

# ============================================================
# SAVE MODELS
# ============================================================

rf_pkl_path = os.path.join(
    OUTPUT_DIR,
    PREFIX + "random_forest_model.pkl"
)

mlp_pkl_path = os.path.join(
    OUTPUT_DIR,
    PREFIX + "mlp_model.pkl"
)

scaler_path = os.path.join(
    OUTPUT_DIR,
    PREFIX + "scaler.pkl"
)

encoder_path = os.path.join(
    OUTPUT_DIR,
    PREFIX + "label_encoders.pkl"
)

selected_features_path = os.path.join(
    OUTPUT_DIR,
    PREFIX + "selected_features.pkl"
)

h5_path = os.path.join(
    OUTPUT_DIR,
    PREFIX + "model_data.h5"
)

joblib.dump(rf_model, rf_pkl_path)
joblib.dump(mlp_model, mlp_pkl_path)
joblib.dump(scaler, scaler_path)
joblib.dump(label_encoders, encoder_path)
joblib.dump(selected_features, selected_features_path)

with h5py.File(h5_path, "w") as h5:

    h5.create_dataset(
        "selected_feature_indices",
        data=selected_indices
    )

    h5.create_dataset(
        "rf_predictions",
        data=rf_pred
    )

    h5.create_dataset(
        "mlp_predictions",
        data=mlp_pred
    )

    h5.create_dataset(
        "actual_values",
        data=y_test.values
    )

    h5.attrs["project_name"] = (
        "Chemical Production Forecasting using AIS + PSO"
    )

# ============================================================
# SAVE CSV
# ============================================================

prediction_df = pd.DataFrame({
    "Actual": y_test.values,
    "PIS_RF_Predicted": rf_pred,
    "PIS_MLP_Predicted": mlp_pred
})

prediction_csv_path = os.path.join(
    OUTPUT_DIR,
    PREFIX + "prediction.csv"
)

prediction_df.to_csv(
    prediction_csv_path,
    index=False
)

results_df = pd.DataFrame({
    "Model": [
        "AIS + PSO + RF",
        "AIS + PSO + MLP"
    ],
    "MAE": [rf_mae, mlp_mae],
    "MSE": [rf_mse, mlp_mse],
    "RMSE": [rf_rmse, mlp_rmse],
    "R2_Score": [rf_r2, mlp_r2],
    "Accuracy_Percentage": [
        rf_r2 * 100,
        mlp_r2 * 100
    ]
})

result_csv_path = os.path.join(
    OUTPUT_DIR,
    PREFIX + "result.csv"
)

results_df.to_csv(
    result_csv_path,
    index=False
)

# ============================================================
# SAVE JSON
# ============================================================

json_report = {
    "project_name": (
        "Chemical Production Forecasting using AIS + PSO"
    ),
    "algorithm": "Hybrid AIS + PSO",
    "target_column": target_col,
    "selected_features": selected_features,
    "best_score": float(pis_best_score)
}

json_path = os.path.join(
    OUTPUT_DIR,
    PREFIX + "report.json"
)

with open(json_path, "w") as f:
    json.dump(json_report, f, indent=4)

# ============================================================
# SAVE YAML
# ============================================================

yaml_config = {
    "project": {
        "name": (
            "Chemical Production Forecasting using AIS + PSO"
        ),
        "author": "Sagnik Patra"
    },
    "paths": {
        "dataset": DATASET_PATH,
        "output_directory": OUTPUT_DIR
    },
    "selected_features": selected_features
}

yaml_path = os.path.join(
    OUTPUT_DIR,
    PREFIX + "config.yaml"
)

with open(yaml_path, "w") as f:
    yaml.dump(yaml_config, f)

# ============================================================
# OPTIMIZATION GRAPH
# ============================================================

optimization_graph_path = os.path.join(
    OUTPUT_DIR,
    PREFIX + "optimization_progress_graph.png"
)

plt.figure(figsize=(9, 5))
plt.plot(pis_progress, marker="o")
plt.title("AIS + PSO Optimization Progress")
plt.xlabel("Generation")
plt.ylabel("Best R2 Score")
plt.tight_layout()
plt.savefig(optimization_graph_path, dpi=300)
plt.show()

# ============================================================
# ACCURACY GRAPH
# ============================================================

accuracy_graph_path = os.path.join(
    OUTPUT_DIR,
    PREFIX + "accuracy_graph.png"
)

plt.figure(figsize=(8, 5))

sns.barplot(
    x="Model",
    y="Accuracy_Percentage",
    data=results_df
)

plt.title("AIS + PSO Accuracy Comparison")
plt.tight_layout()
plt.savefig(accuracy_graph_path, dpi=300)
plt.show()

# ============================================================
# RESULT GRAPH
# ============================================================

result_graph_path = os.path.join(
    OUTPUT_DIR,
    PREFIX + "result_graph.png"
)

plt.figure(figsize=(8, 5))

sns.barplot(
    x="Model",
    y="RMSE",
    data=results_df
)

plt.title("AIS + PSO RMSE Comparison")
plt.tight_layout()
plt.savefig(result_graph_path, dpi=300)
plt.show()

# ============================================================
# COMPARISON GRAPH
# ============================================================

comparison_graph_path = os.path.join(
    OUTPUT_DIR,
    PREFIX + "comparison_graph.png"
)

plt.figure(figsize=(10, 6))

plt.plot(
    y_test.values,
    label="Actual",
    marker="o"
)

plt.plot(
    rf_pred,
    label="RF Prediction",
    marker="x"
)

plt.plot(
    mlp_pred,
    label="MLP Prediction",
    marker="s"
)

plt.title("AIS + PSO Actual vs Predicted")

plt.xlabel("Test Sample")

plt.ylabel(target_col)

plt.legend()

plt.tight_layout()

plt.savefig(comparison_graph_path, dpi=300)

plt.show()

# ============================================================
# PREDICTION GRAPH
# ============================================================

prediction_graph_path = os.path.join(
    OUTPUT_DIR,
    PREFIX + "prediction_graph.png"
)

plt.figure(figsize=(10, 6))

plt.scatter(
    y_test.values,
    rf_pred,
    label="RF"
)

plt.scatter(
    y_test.values,
    mlp_pred,
    label="MLP"
)

plt.plot(
    [y_test.min(), y_test.max()],
    [y_test.min(), y_test.max()],
    linestyle="--"
)

plt.title("AIS + PSO Prediction Graph")

plt.xlabel("Actual")

plt.ylabel("Predicted")

plt.legend()

plt.tight_layout()

plt.savefig(prediction_graph_path, dpi=300)

plt.show()

# ============================================================
# HEATMAP
# ============================================================

heatmap_path = os.path.join(
    OUTPUT_DIR,
    PREFIX + "heatmap.png"
)

encoded_df = df.copy()

for col in encoded_df.columns:
    if encoded_df[col].dtype == "object":

        le = LabelEncoder()

        encoded_df[col] = le.fit_transform(
            encoded_df[col].astype(str)
        )

plt.figure(figsize=(12, 8))

sns.heatmap(
    encoded_df.corr(),
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("AIS + PSO Correlation Heatmap")

plt.tight_layout()

plt.savefig(heatmap_path, dpi=300)

plt.show()

# ============================================================
# TRAINING LOSS GRAPH
# ============================================================

loss_graph_path = os.path.join(
    OUTPUT_DIR,
    PREFIX + "training_loss_graph.png"
)

plt.figure(figsize=(8, 5))

plt.plot(
    mlp_model.loss_curve_,
    label="MLP Training Loss"
)

plt.title("AIS + PSO MLP Training Loss")

plt.xlabel("Iteration")

plt.ylabel("Loss")

plt.legend()

plt.tight_layout()

plt.savefig(loss_graph_path, dpi=300)

plt.show()

# ============================================================
# FINAL OUTPUT
# ============================================================

print("\nAll Hybrid AIS + PSO files saved successfully!")

print("Prediction CSV:", prediction_csv_path)
print("Result CSV:", result_csv_path)
print("JSON:", json_path)
print("YAML:", yaml_path)
print("H5:", h5_path)
print("RF Model:", rf_pkl_path)
print("MLP Model:", mlp_pkl_path)
print("Optimization Graph:", optimization_graph_path)
print("Accuracy Graph:", accuracy_graph_path)
print("Result Graph:", result_graph_path)
print("Comparison Graph:", comparison_graph_path)
print("Prediction Graph:", prediction_graph_path)
print("Heatmap:", heatmap_path)
print("Loss Graph:", loss_graph_path)
```
