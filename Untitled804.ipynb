```python
import os
import json
import yaml
import h5py
import joblib
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

warnings.filterwarnings("ignore")

DATASET_PATH = r"C:\Users\NXTWAVE\Downloads\Chemical Production Forecasting\Production_of_Major_Chemicals_2024-25_1_0.csv"
OUTPUT_DIR = r"C:\Users\NXTWAVE\Downloads\Chemical Production Forecasting"

os.makedirs(OUTPUT_DIR, exist_ok=True)

PREFIX = "ga_"

# ============================================================
# LOAD DATA
# ============================================================

df = pd.read_csv(DATASET_PATH)
df.columns = df.columns.str.strip()

print("Dataset Loaded Successfully")
print("Shape:", df.shape)
print(df.head())

# ============================================================
# CLEAN DATA
# ============================================================

df = df.replace(["-", "NA", "N/A", "na", "n/a", "", " "], np.nan)

for col in df.columns:
    if df[col].dtype == "object":
        cleaned = df[col].astype(str).str.replace(",", "", regex=False).str.strip()
        df[col] = pd.to_numeric(cleaned, errors="ignore")

for col in df.columns:
    if df[col].dtype == "object":
        df[col] = df[col].fillna("Unknown")
    else:
        df[col] = df[col].fillna(df[col].median())

# ============================================================
# TARGET SELECTION
# ============================================================

numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()

if len(numeric_cols) == 0:
    raise ValueError("No numeric column found for prediction.")

target_col = numeric_cols[-1]

print("Target Column Selected:", target_col)

X = df.drop(columns=[target_col])
y = df[target_col]

# ============================================================
# ENCODING
# ============================================================

label_encoders = {}

for col in X.columns:
    if X[col].dtype == "object":
        le = LabelEncoder()
        X[col] = le.fit_transform(X[col].astype(str))
        label_encoders[col] = le

feature_names = list(X.columns)

# ============================================================
# SCALING
# ============================================================

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ============================================================
# TRAIN TEST SPLIT
# ============================================================

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.2,
    random_state=42
)

# ============================================================
# GA FEATURE SELECTION
# Genetic Algorithm
# ============================================================

def evaluate_solution(solution, X_train, X_test, y_train, y_test):
    selected_features = np.where(solution == 1)[0]

    if len(selected_features) == 0:
        return -9999

    model = RandomForestRegressor(
        n_estimators=100,
        max_depth=10,
        random_state=42
    )

    model.fit(X_train[:, selected_features], y_train)
    pred = model.predict(X_test[:, selected_features])

    return r2_score(y_test, pred)


def tournament_selection(population, scores, tournament_size=3):
    selected_indices = np.random.choice(len(population), tournament_size, replace=False)
    best_index = selected_indices[np.argmax(scores[selected_indices])]
    return population[best_index].copy()


def single_point_crossover(parent1, parent2, crossover_rate=0.8):
    if np.random.rand() > crossover_rate:
        return parent1.copy(), parent2.copy()

    if len(parent1) < 2:
        return parent1.copy(), parent2.copy()

    point = np.random.randint(1, len(parent1))

    child1 = np.concatenate([parent1[:point], parent2[point:]])
    child2 = np.concatenate([parent2[:point], parent1[point:]])

    return child1, child2


def mutate(chromosome, mutation_rate=0.1):
    mutation_mask = np.random.rand(len(chromosome)) < mutation_rate
    chromosome[mutation_mask] = 1 - chromosome[mutation_mask]

    if np.sum(chromosome) == 0:
        chromosome[np.random.randint(0, len(chromosome))] = 1

    return chromosome


def ga_feature_selection(
    X_train,
    X_test,
    y_train,
    y_test,
    num_features,
    population_size=30,
    generations=30,
    crossover_rate=0.8,
    mutation_rate=0.1,
    elite_count=2
):
    population = np.random.randint(0, 2, size=(population_size, num_features))

    for i in range(population_size):
        if np.sum(population[i]) == 0:
            population[i, np.random.randint(0, num_features)] = 1

    best_solution = None
    best_score = -9999
    progress = []

    for generation in range(generations):
        scores = np.array([
            evaluate_solution(individual, X_train, X_test, y_train, y_test)
            for individual in population
        ])

        sorted_indices = np.argsort(scores)[::-1]
        population = population[sorted_indices]
        scores = scores[sorted_indices]

        if scores[0] > best_score:
            best_score = scores[0]
            best_solution = population[0].copy()

        progress.append(best_score)

        new_population = []

        elites = population[:elite_count]
        for elite in elites:
            new_population.append(elite.copy())

        while len(new_population) < population_size:
            parent1 = tournament_selection(population, scores)
            parent2 = tournament_selection(population, scores)

            child1, child2 = single_point_crossover(
                parent1,
                parent2,
                crossover_rate
            )

            child1 = mutate(child1, mutation_rate)
            child2 = mutate(child2, mutation_rate)

            new_population.append(child1)

            if len(new_population) < population_size:
                new_population.append(child2)

        population = np.array(new_population[:population_size])

        print(
            f"Generation {generation + 1}/{generations} | "
            f"Best R2 Score: {best_score:.4f} | "
            f"Selected Features: {np.sum(best_solution)}"
        )

    selected_indices = np.where(best_solution == 1)[0]

    return selected_indices, best_score, progress


selected_indices, ga_best_score, ga_progress = ga_feature_selection(
    X_train,
    X_test,
    y_train,
    y_test,
    num_features=X_train.shape[1],
    population_size=30,
    generations=30,
    crossover_rate=0.8,
    mutation_rate=0.1,
    elite_count=2
)

selected_features = [feature_names[i] for i in selected_indices]

print("\nGA Selected Features:")
print(selected_features)

X_train_ga = X_train[:, selected_indices]
X_test_ga = X_test[:, selected_indices]

# ============================================================
# MODEL 1: RANDOM FOREST
# ============================================================

rf_model = RandomForestRegressor(
    n_estimators=200,
    max_depth=10,
    random_state=42
)

rf_model.fit(X_train_ga, y_train)
rf_pred = rf_model.predict(X_test_ga)

rf_mae = mean_absolute_error(y_test, rf_pred)
rf_mse = mean_squared_error(y_test, rf_pred)
rf_rmse = np.sqrt(rf_mse)
rf_r2 = r2_score(y_test, rf_pred)

# ============================================================
# MODEL 2: MLP REGRESSOR
# TensorFlow removed to avoid DLL paging file error
# ============================================================

mlp_model = MLPRegressor(
    hidden_layer_sizes=(128, 64, 32),
    activation="relu",
    solver="adam",
    max_iter=1000,
    random_state=42,
    early_stopping=True
)

mlp_model.fit(X_train_ga, y_train)
mlp_pred = mlp_model.predict(X_test_ga)

mlp_mae = mean_absolute_error(y_test, mlp_pred)
mlp_mse = mean_squared_error(y_test, mlp_pred)
mlp_rmse = np.sqrt(mlp_mse)
mlp_r2 = r2_score(y_test, mlp_pred)

# ============================================================
# SAVE MODELS
# ============================================================

rf_pkl_path = os.path.join(OUTPUT_DIR, PREFIX + "random_forest_model.pkl")
mlp_pkl_path = os.path.join(OUTPUT_DIR, PREFIX + "mlp_model.pkl")
scaler_path = os.path.join(OUTPUT_DIR, PREFIX + "scaler.pkl")
encoder_path = os.path.join(OUTPUT_DIR, PREFIX + "label_encoders.pkl")
selected_features_path = os.path.join(OUTPUT_DIR, PREFIX + "selected_features.pkl")
h5_path = os.path.join(OUTPUT_DIR, PREFIX + "model_data.h5")

joblib.dump(rf_model, rf_pkl_path)
joblib.dump(mlp_model, mlp_pkl_path)
joblib.dump(scaler, scaler_path)
joblib.dump(label_encoders, encoder_path)
joblib.dump(selected_features, selected_features_path)

with h5py.File(h5_path, "w") as h5:
    h5.create_dataset("selected_feature_indices", data=selected_indices)
    h5.create_dataset("rf_predictions", data=rf_pred)
    h5.create_dataset("mlp_predictions", data=mlp_pred)
    h5.create_dataset("actual_values", data=y_test.values)
    h5.attrs["project_name"] = "Chemical Production Forecasting using GA"
    h5.attrs["target_column"] = target_col
    h5.attrs["ga_best_score"] = ga_best_score

# ============================================================
# SAVE CSV FILES
# ============================================================

prediction_df = pd.DataFrame({
    "Actual": y_test.values,
    "GA_Random_Forest_Predicted": rf_pred,
    "GA_MLP_Predicted": mlp_pred
})

prediction_csv_path = os.path.join(OUTPUT_DIR, PREFIX + "prediction.csv")
prediction_df.to_csv(prediction_csv_path, index=False)

results_df = pd.DataFrame({
    "Model": ["GA + Random Forest", "GA + MLP Regressor"],
    "MAE": [rf_mae, mlp_mae],
    "MSE": [rf_mse, mlp_mse],
    "RMSE": [rf_rmse, mlp_rmse],
    "R2_Score": [rf_r2, mlp_r2],
    "Accuracy_Percentage": [rf_r2 * 100, mlp_r2 * 100],
    "Selected_Features_Count": [len(selected_features), len(selected_features)]
})

result_csv_path = os.path.join(OUTPUT_DIR, PREFIX + "result.csv")
results_df.to_csv(result_csv_path, index=False)

print("\nGA Result CSV:")
print(results_df)

# ============================================================
# JSON REPORT
# ============================================================

json_report = {
    "project_name": "Chemical Production Forecasting System using GA",
    "author": "Sagnik Patra",
    "algorithm": "Genetic Algorithm Feature Selection",
    "note": "TensorFlow removed. MLPRegressor used to avoid DLL paging file error.",
    "target_column": target_col,
    "rows": int(df.shape[0]),
    "columns": int(df.shape[1]),
    "total_features": int(len(feature_names)),
    "selected_features_count": int(len(selected_features)),
    "selected_features": selected_features,
    "ga_best_score": float(ga_best_score),
    "random_forest": {
        "mae": float(rf_mae),
        "mse": float(rf_mse),
        "rmse": float(rf_rmse),
        "r2_score": float(rf_r2),
        "accuracy_percentage": float(rf_r2 * 100)
    },
    "mlp_regressor": {
        "mae": float(mlp_mae),
        "mse": float(mlp_mse),
        "rmse": float(mlp_rmse),
        "r2_score": float(mlp_r2),
        "accuracy_percentage": float(mlp_r2 * 100)
    }
}

json_path = os.path.join(OUTPUT_DIR, PREFIX + "report.json")

with open(json_path, "w") as f:
    json.dump(json_report, f, indent=4)

# ============================================================
# YAML CONFIG
# ============================================================

yaml_config = {
    "project": {
        "name": "Chemical Production Forecasting System using GA",
        "author": "Sagnik Patra"
    },
    "note": "TensorFlow removed due to DLL paging file error. MLPRegressor used instead.",
    "paths": {
        "dataset_path": DATASET_PATH,
        "output_directory": OUTPUT_DIR,
        "h5_file": h5_path,
        "rf_pkl_model": rf_pkl_path,
        "mlp_pkl_model": mlp_pkl_path,
        "prediction_csv": prediction_csv_path,
        "result_csv": result_csv_path,
        "json_report": json_path
    },
    "data": {
        "target_column": target_col,
        "rows": int(df.shape[0]),
        "columns": int(df.shape[1]),
        "all_features": feature_names,
        "selected_features": selected_features
    },
    "ga_parameters": {
        "population_size": 30,
        "generations": 30,
        "crossover_rate": 0.8,
        "mutation_rate": 0.1,
        "elite_count": 2
    },
    "models": {
        "random_forest": {
            "n_estimators": 200,
            "max_depth": 10,
            "random_state": 42
        },
        "mlp_regressor": {
            "hidden_layer_sizes": [128, 64, 32],
            "activation": "relu",
            "solver": "adam",
            "max_iter": 1000,
            "early_stopping": True
        }
    }
}

yaml_path = os.path.join(OUTPUT_DIR, PREFIX + "config.yaml")

with open(yaml_path, "w") as f:
    yaml.dump(yaml_config, f, sort_keys=False)

# ============================================================
# GRAPHS
# ============================================================

ga_progress_graph_path = os.path.join(
    OUTPUT_DIR,
    PREFIX + "optimization_progress_graph.png"
)

plt.figure(figsize=(9, 5))
plt.plot(ga_progress, marker="o")
plt.title("GA Optimization Progress Graph")
plt.xlabel("Generation")
plt.ylabel("Best R2 Score")
plt.tight_layout()
plt.savefig(ga_progress_graph_path, dpi=300)
plt.show()

accuracy_graph_path = os.path.join(OUTPUT_DIR, PREFIX + "accuracy_graph.png")

plt.figure(figsize=(8, 5))
sns.barplot(x="Model", y="Accuracy_Percentage", data=results_df)
plt.title("GA Model Accuracy Comparison")
plt.xlabel("Model")
plt.ylabel("Accuracy Percentage")
plt.xticks(rotation=15)
plt.tight_layout()
plt.savefig(accuracy_graph_path, dpi=300)
plt.show()

result_graph_path = os.path.join(OUTPUT_DIR, PREFIX + "result_graph.png")

plt.figure(figsize=(8, 5))
sns.barplot(x="Model", y="RMSE", data=results_df)
plt.title("GA Model RMSE Result Comparison")
plt.xlabel("Model")
plt.ylabel("RMSE")
plt.xticks(rotation=15)
plt.tight_layout()
plt.savefig(result_graph_path, dpi=300)
plt.show()

comparison_graph_path = os.path.join(OUTPUT_DIR, PREFIX + "comparison_graph.png")

plt.figure(figsize=(10, 6))
plt.plot(y_test.values, label="Actual", marker="o")
plt.plot(rf_pred, label="GA Random Forest Predicted", marker="x")
plt.plot(mlp_pred, label="GA MLP Predicted", marker="s")
plt.title("GA Actual vs Predicted Comparison")
plt.xlabel("Test Sample Index")
plt.ylabel(target_col)
plt.legend()
plt.tight_layout()
plt.savefig(comparison_graph_path, dpi=300)
plt.show()

prediction_graph_path = os.path.join(OUTPUT_DIR, PREFIX + "prediction_graph.png")

plt.figure(figsize=(10, 6))
plt.scatter(y_test.values, rf_pred, label="GA Random Forest")
plt.scatter(y_test.values, mlp_pred, label="GA MLP")
plt.plot(
    [y_test.min(), y_test.max()],
    [y_test.min(), y_test.max()],
    linestyle="--",
    label="Perfect Prediction Line"
)
plt.title("GA Prediction Graph")
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.legend()
plt.tight_layout()
plt.savefig(prediction_graph_path, dpi=300)
plt.show()

heatmap_path = os.path.join(OUTPUT_DIR, PREFIX + "heatmap.png")

encoded_df = df.copy()

for col in encoded_df.columns:
    if encoded_df[col].dtype == "object":
        le = LabelEncoder()
        encoded_df[col] = le.fit_transform(encoded_df[col].astype(str))

plt.figure(figsize=(12, 8))
sns.heatmap(encoded_df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("GA Feature Correlation Heatmap")
plt.tight_layout()
plt.savefig(heatmap_path, dpi=300)
plt.show()

loss_graph_path = os.path.join(OUTPUT_DIR, PREFIX + "training_loss_graph.png")

plt.figure(figsize=(8, 5))
plt.plot(mlp_model.loss_curve_, label="MLP Training Loss")
plt.title("GA MLP Training Loss Graph")
plt.xlabel("Iteration")
plt.ylabel("Loss")
plt.legend()
plt.tight_layout()
plt.savefig(loss_graph_path, dpi=300)
plt.show()

# ============================================================
# FINAL OUTPUT
# ============================================================

print("\nAll GA files saved successfully!")

print("RF PKL Model:", rf_pkl_path)
print("MLP PKL Model:", mlp_pkl_path)
print("H5 File:", h5_path)
print("Scaler:", scaler_path)
print("Encoders:", encoder_path)
print("Selected Features:", selected_features_path)
print("YAML Config:", yaml_path)
print("JSON Report:", json_path)
print("Result CSV:", result_csv_path)
print("Prediction CSV:", prediction_csv_path)
print("Optimization Graph:", ga_progress_graph_path)
print("Accuracy Graph:", accuracy_graph_path)
print("Result Graph:", result_graph_path)
print("Comparison Graph:", comparison_graph_path)
print("Prediction Graph:", prediction_graph_path)
print("Heatmap:", heatmap_path)
print("Training Loss Graph:", loss_graph_path)
```
